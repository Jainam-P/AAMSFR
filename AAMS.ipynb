```python
if not os.path.exists('Teachers'):
    os.makedirs('Teachers')

global root
root = tk.Tk()

root.title("login page")
root.iconbitmap('images/icons/i.ico')

root.state('zoomed')
style = ttk.Style()
mygreen = "#d2ffd2"
myred = "#dd0202"

style.theme_create( "yummy", parent="alt", settings={
        
        "TNotebook.Tab": {
            "configure": {"background": "black" },
             "map":       {"background": [("selected", "#24f2e8")],"foreground": [("selected", "black")]}  } } )

style.theme_use("yummy")
style.theme_use("yummy")
style.configure('TNotebook.Tab', font=('Calibri', '18','bold'),foreground="white",borderwidth=0, padding=[300, 5])

s = ttk.Style()
s.configure('My.TFrame', background="#000000",activebackground='#000000')

def delete_login_success():
    login_success_screen.destroy()

def delete_password_not_recognised():
    password_not_recog_screen.destroy()

def delete_user_not_found_screen():
    user_not_found_screen.destroy()

def login():
    global userName
    userName = Uname.get()
    passWord = Pword.get()
    Uentry.delete(0, END)
    Pentry.delete(0, END)
    if not os.path.exists('Teachers/'+userName+'/'+userName):
        user_not_found()
    list_of_files = os.listdir('Teachers')
    if userName in list_of_files:
        file1 = open('Teachers/'+userName+'/'+userName, "r")
        verify = file1.read().splitlines()

        if passWord in verify:
            login_sucess()

        else:
            password_not_recognised()

def login_sucess():
    global login_success_screen
    login_success_screen = Toplevel(root)
    login_success_screen.title("Success")
    login_success_screen.geometry("300x100")
    LAB2=Label(login_success_screen, text="login_success",width=30,bg="black", fg="RED", font=("calibri", 20)).pack()
    Button(login_success_screen, text="OK", command=delete_login_success).pack()
    root.destroy()
# Designing popup for login invalid password

def password_not_recognised():
    global password_not_recog_screen
    password_not_recog_screen = Toplevel(root)
    password_not_recog_screen.title("Unsuccess")
    password_not_recog_screen.geometry("300x100")
    LAB2=Label(password_not_recog_screen, text="password not recognized",width=30,bg="black", fg="RED", font=("calibri", 20)).pack()
    Button(password_not_recog_screen, text="OK", command=delete_password_not_recognised).pack()

# Designing popup for user not found

def user_not_found():
    global user_not_found_screen
    user_not_found_screen = Toplevel(root)
    user_not_found_screen.title("user_not_found_screen")
    user_not_found_screen.geometry("300x100")
    LAB2=Label(user_not_found_screen, text="user not found ",width=30,bg="black", fg="RED", font=("calibri", 20)).pack()
    Button(user_not_found_screen, text="OK", command=delete_user_not_found_screen).pack()

def register():
        
    global register_success_screen
    register_success_screen = Toplevel(root)
    Username = uname.get()
    Emailid = eid.get()
    Password = pword.get()
    Cpassword = cword.get()
    if Password == Cpassword:
        if os.path.exists('Teachers/'+Username):
            Label(register_success_screen, text="You are alredy registered ", fg="green", font=("calibri", 20)).grid(row=7,column=2,padx=150, pady=30)

        if not os.path.exists('Teachers/'+Username):
            os.makedirs('Teachers/'+Username)
            file = open('Teachers/'+Username+'/'+Username, "w")
            file.write(Username + "\n")
            file.write(Emailid + "\n")
            file.write(Password +"\n")
            file.close()
            Label(register_success_screen, text="Registration Success", fg="green", font=("calibri", 20)).grid(row=7,column=2,padx=150, pady=30)
    else:
        Label(register_success_screen, text="Password and Confirm password must be same", fg="green", font=("calibri", 20)).grid(row=7,column=2,padx=150, pady=30)

    uEntry.delete(0, END)
    eEntry.delete(0, END)
    pEntry.delete(0, END)
    cEntry.delete(0, END)

def doSomething():
    root.destroy()
    exit()

root.protocol('WM_DELETE_WINDOW', doSomething)  # root is your root window

tab_parent = ttk.Notebook(root)

tab1 = ttk.Frame(tab_parent)
img1 = PhotoImage(file="images/bg/slide1.png")
BGlabel = Label(tab1, image=img1)
BGlabel.place(x=0, y=0, relwidth=1, relheight=1)

tab2 = ttk.Frame(tab_parent)
img2 = PhotoImage(file="images/bg/slide1.png")
BGlabel = Label(tab2, image=img2)
BGlabel.place(x=0, y=0, relwidth=1, relheight=1)

tab_parent.add(tab1, text=" LOGIN ")
tab_parent.add(tab2, text=" SIGN UP ")
tab_parent.pack(expand=True, fill=tk.BOTH)

def select_reg():
    tab_parent.select(1)

def select_log():
    tab_parent.select(0)

global Uname
global Pword
Uname = StringVar()
Pword = StringVar()

Ftab3=Frame(tab1,bg="#24f2e8")
Ftab3.place(x=50,y=80,height=500,width=268)
Label(Ftab3, text="WELCOME BACK !",fg="black",bg="#24f2e8",font=("Calibri", 24,'bold')).place(x=20,y=200)
Label(Ftab3, text="To The Automated Attendance ",fg="black",bg="#24f2e8",font=("Calibri", 12)).place(x=40,y=250)
Label(Ftab3, text="Management System A Machine",fg="black",bg="#24f2e8",font=("Calibri", 12)).place(x=30,y=270)
Label(Ftab3, text="Learning Based Software  ",fg="black",bg="#24f2e8",font=("Calibri", 12)).place(x=50,y=290)
Label(Ftab3, text="Are You New User ? ",fg="black",bg="#24f2e8",font=("Calibri", 12)).place(x=70,y=330)

def Ftab3RI(e):
    Ftab3Ftab3RImage=PhotoImage(file="images/buttons/bu1.png")
    Ftab3Rbutton.config(image=Ftab3Ftab3RImage)
    Ftab3Rbutton.image=Ftab3Ftab3RImage
def Ftab3ri(e):
    Ftab3Ftab3RImage=PhotoImage(file="images/buttons/bu.png")
    Ftab3Rbutton.config(image=Ftab3Ftab3RImage)
    Ftab3Rbutton.image=Ftab3Ftab3RImage
Ftab3Ftab3RImage=PhotoImage(file="images/buttons/bu.png")
Ftab3Rbutton = tk.Button(Ftab3, image=Ftab3Ftab3RImage,bg="#000000", borderwidth=0,command=select_reg)
Ftab3Rbutton.bind("<Enter>", Ftab3RI)
Ftab3Rbutton.bind("<Leave>", Ftab3ri)

Ftab3Rbutton.place(x=50,y=380)

Ftab1=Frame(tab1,bg="#000000")
Ftab1.place(x=320,y=80,height=500,width=400)
Label(Ftab1, text="PLEASE ENTER YOUR USERNAME",fg="#24f2e8",bg="#000000",font=("Calibri", 18)).place(x=30,y=30)
Label(Ftab1, text="AND PASSWORD",fg="#24f2e8",bg="#000000",font=("Calibri", 18)).place(x=30,y=80)
Label(Ftab1, text="Teacher Username :",fg="#FFFFFF",font=("Calibri", 16), borderwidth=0,bg="#000000").place(x=30,y=160)
Label(Ftab1, text="Password : ",fg="#FFFFFF",font=("Calibri", 16), borderwidth=0,bg="#000000").place(x=30,y=260)

Uentry=Entry(Ftab1, textvariable=Uname,font=("Calibri", 16),fg="#00FFFF",bg="#000000", borderwidth=1,insertbackground="#00FFFF")
Uentry.place(x=30,y=210)
Pentry=Entry(Ftab1, textvariable=Pword,show="*",font=("Calibri", 16),fg="#00FFFF",bg="#000000", borderwidth=1,insertbackground="#00FFFF")
Pentry.place(x=30,y=310)

def clr():
    Uentry.delete(0, END)
    Pentry.delete(0, END)

def LI(e):
    Limage=PhotoImage(file="images/buttons/l.png")
    Lbutton.config(image=Limage)
    Lbutton.image=Limage
def li(e):
    Limage=PhotoImage(file="images/buttons/l1.png")
    Lbutton.config(image=Limage)
    Lbutton.image=Limage
Limage=PhotoImage(file="images/buttons/l1.png")
Lbutton = tk.Button(tab1, image=Limage,bg="#000000", borderwidth=0,command=login)
Lbutton.bind("<Enter>", LI)
Lbutton.bind("<Leave>", li)

def RI(e):
    Rimage=PhotoImage(file="images/buttons/c21.png")
    Rbutton.config(image=Rimage)
    Rbutton.image=Rimage
def ri(e):
    Rimage=PhotoImage(file="images/buttons/c22.png")
    Rbutton.config(image=Rimage)
    Rbutton.image=Rimage
Rimage=PhotoImage(file="images/buttons/c22.png")
Rbutton = tk.Button(tab1, image=Rimage,bg="#000000", borderwidth=0,command=clr)
Rbutton.bind("<Enter>", RI)
Rbutton.bind("<Leave>", ri)

Lbutton.place(x=540,y=500)

Rbutton.place(x=340,y=500)

t2imagei=PhotoImage(file="images/icons/2.png")
t2buttoni = tk.Button(tab1, image=t2imagei,bg="#24f2e8",activebackground='black', borderwidth=0)
t2buttoni.place(x=110,y=100)

##################################################################################################################################
global uname
global pword
global eid
global cword
uname = StringVar()
pword = StringVar()
eid = StringVar()
cword = StringVar()

Ftab4=Frame(tab2,bg="#24f2e8")
Ftab4.place(x=500,y=80,height=500,width=268)
Label(Ftab4, text="HELLO THERE !",fg="black",bg="#24f2e8",font=("Calibri", 24,'bold')).place(x=40,y=200)
Label(Ftab4, text="Welcome To The Automated  ",fg="black",bg="#24f2e8",font=("Calibri", 12)).place(x=45,y=250)
Label(Ftab4, text="Attendance Management System ",fg="black",bg="#24f2e8",font=("Calibri", 12)).place(x=30,y=275)
Label(Ftab4, text="A Machine Learning Based Software  ",fg="black",bg="#24f2e8",font=("Calibri", 12)).place(x=20,y=300)
Label(Ftab4, text="You Can Login From Here",fg="black",bg="#24f2e8",font=("Calibri", 12)).place(x=40,y=325)
Label(Ftab4, text="After Sign UP ",fg="black",bg="#24f2e8",font=("Calibri", 12)).place(x=80,y=350)

def Ftab4RI(e):
    Ftab4Ftab4RImage=PhotoImage(file="images/buttons/l1.png")
    Ftab4Rbutton.config(image=Ftab4Ftab4RImage)
    Ftab4Rbutton.image=Ftab4Ftab4RImage
def Ftab4ri(e):
    Ftab4Ftab4RImage=PhotoImage(file="images/buttons/l.png")
    Ftab4Rbutton.config(image=Ftab4Ftab4RImage)
    Ftab4Rbutton.image=Ftab4Ftab4RImage
Ftab4Ftab4RImage=PhotoImage(file="images/buttons/l.png")
Ftab4Rbutton = tk.Button(Ftab4, image=Ftab4Ftab4RImage,bg="#000000", borderwidth=0,command=select_log)
Ftab4Rbutton.bind("<Enter>", Ftab4RI)
Ftab4Rbutton.bind("<Leave>", Ftab4ri)

Ftab4Rbutton.place(x=55,y=400)

Ftab2=Frame(tab2,bg="#000000")
Ftab2.place(x=50,y=80,height=500,width=450)
Label(Ftab2, text="PLEASE ENTER YOUR DETAILS ",fg="#00FFFF",bg="#000000",font=("Calibri", 20)).place(x=60,y=20)

Label(Ftab2, text="Username :",fg="#FFFFFF",font=("Calibri", 15), borderwidth=0,bg="#000000").place(x=30,y=100)
Label(Ftab2, text="Email ID :",fg="#FFFFFF",font=("Calibri", 15), borderwidth=0,bg="#000000").place(x=30,y=180)
Label(Ftab2, text="Password :",fg="#FFFFFF",font=("Calibri", 15), borderwidth=0,bg="#000000").place(x=30,y=260)
Label(Ftab2, text="Confirm Password :",fg="#FFFFFF",font=("Calibri", 15), borderwidth=0,bg="#000000").place(x=30,y=340)

uEntry=Entry(Ftab2, textvariable=uname,font=("Calibri", 15),fg="#00FFFF",bg="#000000", borderwidth=1,insertbackground="#00FFFF")
uEntry.place(x=200,y=100)
eEntry=Entry(Ftab2, textvariable=eid,font=("Calibri", 15),fg="#00FFFF",bg="#000000", borderwidth=1,insertbackground="#00FFFF")
eEntry.place(x=200,y=180)
pEntry=Entry(Ftab2, textvariable=pword,show="*",font=("Calibri", 15),fg="#00FFFF",bg="#000000", borderwidth=1,insertbackground="#00FFFF")
pEntry.place(x=200,y=260)
cEntry=Entry(Ftab2, textvariable=cword,show="*",font=("Calibri", 15),fg="#00FFFF",bg="#000000", borderwidth=1,insertbackground="#00FFFF")
cEntry.place(x=200,y=340)

def clr2():
    uEntry.delete(0, END)
    eEntry.delete(0, END)
    pEntry.delete(0, END)
    cEntry.delete(0, END)

def IR(e):
    imageR=PhotoImage(file="images/buttons/c21.png")
    buttonR.config(image=imageR)
    buttonR.image=imageR
def ir(e):
    imageR=PhotoImage(file="images/buttons/c22.png")
    buttonR.config(image=imageR)
    buttonR.image=imageR
imageR=PhotoImage(file="images/buttons/c22.png")
buttonR = tk.Button(tab2, image=imageR,bg="#000000", borderwidth=0,command=clr2)
buttonR.bind("<Enter>", IR)
buttonR.bind("<Leave>", ir)

def IL(e):
    imageL=PhotoImage(file="images/buttons/bu.png")
    buttonL.config(image=imageL)
    buttonL.image=imageL
def il(e):
    imageL=PhotoImage(file="images/buttons/b.png")
    buttonL.config(image=imageL)
    buttonL.image=imageL
imageL=PhotoImage(file="images/buttons/b.png")
buttonL = tk.Button(tab2, image=imageL,bg="#000000", borderwidth=0,command=register)
buttonL.bind("<Enter>", IL)
buttonL.bind("<Leave>", il)

buttonR.place(x=80,y=500)
buttonL.place(x=300,y=500)

t2imagei2=PhotoImage(file="images/icons/2.png")
t2buttoni2 = tk.Button(tab2, image=t2imagei2,bg="#24f2e8",activebackground='black', borderwidth=0)
t2buttoni2.place(x=550,y=100)

tab_parent.pack(expand=1, fill='both')

root.mainloop()

#########################################################################################################################
##############################################################
##################################################

#########################################################################################################################
#################     tabs Screen    #############################################
##################################################

#defining A folder
if not os.path.exists('STUDENT_DETAILS'):
    os.makedirs('STUDENT_DETAILS')

#if not os.path.exists('SUBJECT_DETAILS'):
    #os.makedirs('SUBJECT_DETAILS')

form = tk.Tk()
form.title("Attendance System")
form.iconbitmap('images/icons/i.ico')
form.state('zoomed')

#global filename

style = ttk.Style()
style.theme_create( "yummy", parent="alt", settings={
        
        "TNotebook.Tab": {
            "configure": {"background": "black" },
            "map":       {"background": [("selected", "#24f2e8")],"foreground": [("selected", "black")]}  } } )

style.theme_use("yummy")
style.configure('TNotebook.Tab', font=('Calibri', '12'),foreground="white",borderwidth=0, padding=[20, 10])

s = ttk.Style()
h = ttk.Style()
s.configure('My.TFrame', background="black",activebackground='black')
h.configure('home.TFrame', background="black",activebackground='black')

tab_parent = ttk.Notebook(form)

Home = ttk.Frame(tab_parent,style='home.TFrame')
tab1 = ttk.Frame(tab_parent,style='My.TFrame')

#tab1.wm_attributes("-transparentcolor", "black")
global filename
filename = PhotoImage(file = "images/bg/ho.png")
background_label = Label(tab1, image=filename)
background_label.place(x=0, y=0, relwidth=1, relheight=1)

tab2 = ttk.Frame(tab_parent,style='My.TFrame')
tab3 = ttk.Frame(tab_parent,style='My.TFrame')
tab4 = ttk.Frame(tab_parent,style='My.TFrame')
tab5 = ttk.Frame(tab_parent,style='My.TFrame')
tab6 = ttk.Frame(tab_parent,style='My.TFrame')
tab7 = ttk.Frame(tab_parent,style='My.TFrame')
tab8 = ttk.Frame(tab_parent,style='My.TFrame')

#background_label = Label(tab2, image=filename)
#background_label.place(x=0, y=0, relwidth=1, relheight=1)
tab_parent.add(Home, text=" HOME ")

tab_parent.add(tab1, text="STUDENT REGISTRATION ")
tab_parent.add(tab2, text="STUDENT RECORD")
tab_parent.add(tab3, text="COURSE REGISTRATION ")
tab_parent.add(tab4, text="COURSE RECORD")

tab_parent.add(tab5, text="TAKE ATTENDACE")
tab_parent.add(tab6, text="DEFAULTERS ")
tab_parent.add(tab7, text="DETENTION LIST")
tab_parent.add(tab8, text="My Profile")

tab_parent.pack(expand=True, fill=tk.BOTH)

#### tab functions ##############################################################################################################################

global assure_path_exists
def assure_path_exists(path):
    dir = os.path.dirname(path)
    if not os.path.exists(dir):
        os.makedirs(dir)

######## train function #########

def train():
    
    Year_info=selection
    
    if Year_info == '1':
        Yr = 'FY'

    elif Year_info == '2':
            Yr = 'SY'

    elif Year_info == '3':
                Yr = 'TY'
                
    # Create Local Binary Patterns Histograms for face recognization
    recognizer = cv2.face.LBPHFaceRecognizer_create()

    # Using prebuilt frontal face training model, for face detection
    detector = cv2.CascadeClassifier("haarcascade_frontalface_default.xml");

    # Create method to get the images and label data
    def getImagesAndLabels(path):

        # Get all file path
        imagePaths = [os.path.join(path,f) for f in os.listdir(path)] 
        
        # Initialize empty face sample
        faceSamples=[]
        
        # Initialize empty id
        ids = []

        # Loop all the file path
        for imagePath in imagePaths:

            # Get the image and convert it to grayscale
            PIL_img = Image.open(imagePath).convert('L')

            # PIL image to numpy array
            img_numpy = np.array(PIL_img,'uint8')

            # Get the image id
            id = int(os.path.split(imagePath)[-1].split(".")[0])

            # Get the face from the training images
            faces = detector.detectMultiScale(img_numpy)

            # Loop for each face, append to their respective ID
            for (x,y,w,h) in faces:

                # Add the image to face samples
                faceSamples.append(img_numpy[y:y+h,x:x+w])

                # Add the ID to IDs
                ids.append(id)

        # Pass the face array and IDs array
        return faceSamples,ids

    # Get the faces and IDs
    faces,ids = getImagesAndLabels('dataset/'+Yr)

    # Train the model using the faces and IDs
    recognizer.train(faces, np.array(ids))

    # Save the model into trainer.yml
    assure_path_exists('trainer/')
    recognizer.save('trainer/'+Yr+'.yml')
    ##print("done")

######### createdata funtion ################

def createdata():

    global std_id
    std_id = ENR.get()

    global face_name   
    face_name = Sname.get()
 
    Year_info=selection

    if Year_info == '1':
        Yr = 'FY'

    elif Year_info == '2':
            Yr = 'SY'

    elif Year_info == '3':
                Yr = 'TY'
                
    ids = []

    if not os.path.exists('dataset/'+Yr):
        os.makedirs('dataset/'+Yr)
        
    list_of_files = os.listdir('dataset/'+Yr)
    for std in list_of_files :
        id = int(os.path.split(std)[-1].split(".")[0])
        ids.append(id)

    global face_id
    if len(ids) == 0:
        face_id = 1
    else:
        face_id =1+max(ids)
    
    # Start capturing video
    vid_cam = cv2.VideoCapture(0)

    # Detect object in video stream using Haarcascade Frontal Face
    face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

    # Initialize sample face image
    count = 0

    assure_path_exists('dataset/'+Yr)

    #fc=[]
    # Start looping
    while(True):

        # Capture video frame
        _, image_frame = vid_cam.read()

        # Convert frame to grayscale
        gray = cv2.cvtColor(image_frame, cv2.COLOR_BGR2GRAY)

        # Detect frames of different sizes, list of faces rectangles
        faces = face_detector.detectMultiScale(gray, 1.3, 5)

        # Loops for each faces
        for (x,y,w,h) in faces:

            # Crop the image frame into rectangle
            cv2.rectangle(image_frame, (x,y), (x+w,y+h),(0,0,255), 1)

            # Increment sample face image
            count += 1

            # Save the captured image into the datasets folder
            cv2.imwrite('dataset/'+Yr+'/'+str(face_id) + '.' +str(face_name)+ '.' + str(std_id) + '.' + str(count) + ".jpg", gray[y:y+h,x:x+w])

            # Display the video frame, with bounded rectangle on the person's face
            #cv2.imshow('frame', image_frame)

            cv2.namedWindow ('screen', cv2.WINDOW_NORMAL)
            cv2.setWindowProperty ('screen', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)
            cv2.imshow ('screen', image_frame)
            
        # To stop taking video, press 'q' for at least 100ms
        if cv2.waitKey(50) & 0xFF == ord('q'):
            break
        
        # If image taken reach 20, stop taking video
        elif count>50:
            break
        
    # Stop video

    vid_cam.release()
    cv2.destroyAllWindows()
    train()

################## student registration tab1 funtion ####################################################################

def register_student():

    Sname_info = Sname.get()
    ENR_info = ENR.get()
    Email_info = Email.get()
    Phone_info = Phone.get()
    PPhone_info = PPhone.get()
    PEmail_info = PEmail.get()
    Sem_info =SM.get()
    ##print("in function "+selection)
    Year_info=selection
    ##print(Year_info)

    global Yr
    if Year_info == '1':
        Yr = 'FY'
        ##print('assigned 1')

    elif Year_info == '2':
        Yr = 'SY'
        ##print('assigned 2')

    elif Year_info == '3':
        Yr = 'TY'
        ##print('assigned 3')

    ##print(Yr)

    if not os.path.exists('STUDENT_DETAILS/'+Yr):
        os.makedirs('STUDENT_DETAILS/'+Yr)
    
    if os.path.exists('STUDENT_DETAILS/'+Yr+'/'+ENR_info):
        global lb
        lb=Label(tab1, text="",bg="black", fg="#00FFFF", font=("calibri", 20)).place(x=550,y=620)
        lb=Label(tab1, text="The Enrollment already registered",bg="black", fg="#00FFFF", font=("calibri", 20)).place(x=550,y=620)
        Sname_entry.delete(0, END)
        ENR_entry.delete(0, END)
        Email_entry.delete(0, END)
        Phone_entry.delete(0, END)
        SEM_entry.delete(0, END)

    if not os.path.exists('STUDENT_DETAILS/'+Yr+'/'+ENR_info):
        file = open('STUDENT_DETAILS/'+Yr+'/'+ENR_info, "w")
        file.write(Sname_info + "\n")
        file.write(ENR_info + "\n")
        file.write(Email_info + "\n")
        file.write(Phone_info + "\n")
        file.write(Year_info + "\n")
        file.write(Sem_info + "\n")
        file.write(PEmail_info + "\n")
        file.write(PPhone_info + "\n")
        file.close()
        lb=Label(tab1, text="",bg="black", fg="#00FFFF", font=("calibri", 20)).place(x=550,y=620)
        lb=Label(tab1, text="Registration Success",bg="black", fg="#00FFFF", font=("calibri", 20)).place(x=550,y=620)
        createdata()
        Sname_entry.delete(0, END)
        ENR_entry.delete(0, END)
        Email_entry.delete(0, END)
        Phone_entry.delete(0, END)
        SEM_entry.delete(0, END)

def tab1clr():
    Sname_entry.delete(0, END)
    ENR_entry.delete(0, END)
    Email_entry.delete(0, END)
    Phone_entry.delete(0, END)
    SEM_entry.delete(0, END)

######################## tab2 functions #############################################################################
    
def tab2clr():
    DELENR_entry.delete(0, END)
    
def del_student():
    DELENR_info = ENRDEL.get()
    Year_info=tab2selection
    if Year_info == '1':
        Yr = 'FY'
        ##print('assigned 1')

    elif Year_info == '2':
        Yr = 'SY'
        ##print('assigned 2')

    elif Year_info == '3':
        Yr = 'TY'
        ##print('assigned 3')

    if os.path.exists('STUDENT_DETAILS/'+Yr+'/'+DELENR_info):
        SDEL=Frame(tab2,bg="black",highlightbackground="orange",highlightcolor='#24f2e8', highlightthickness=2)
        SDEL.place(x=750,y=10,height=300,width=600)
        tk.Label(SDEL, text="THE STUDENT DETAILS ",fg="#00ffcc",bg="black",font=("Calibri", 16)).place(x=20,y=5)
        tk.Label(SDEL, text="THE PARENT DETAILS ",fg="#00ffcc",bg="black",font=("Calibri", 16)).place(x=320,y=5)

        SDELVR=Frame(SDEL,bg="red")
        SDELVR.place(x=300,y=10,height=190,width=2)
        
        #global ENRDEL
        #ENRDEL = StringVar()
        global det
        det=[]
        with open('STUDENT_DETAILS/'+Yr+'/'+DELENR_info, 'r') as filehandle:
            for line in filehandle:
                det.append(line)
                ##print(line)
        ##print(det)
        newdet = [x[:-1] for x in det]
        print(newdet)
        Label(SDEL, text="NAME ",bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=20,y=60)
        Label(SDEL, text=":  "+newdet[0],bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=120,y=60)
        Label(SDEL, text="ENR NO ",bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=20,y=100)
        Label(SDEL, text=":  "+newdet[1],bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=120,y=100)
        Label(SDEL, text="EMAIL ID ",bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=20,y=140)
        Label(SDEL, text=":  "+newdet[2],bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=120,y=140)
        Label(SDEL, text="PHONE NO ",bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=20,y=180)
        Label(SDEL, text=":  "+newdet[3],bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=120,y=180)
        Label(SDEL, text="EMAIL ID ",bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=320,y=100)
        Label(SDEL, text=":  "+newdet[6],bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=420,y=100)
        Label(SDEL, text="PHONE NO ",bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=320,y=140)
        Label(SDEL, text=":  "+newdet[7],bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=420,y=140)

        tab2stdavg()
        #DELENR_entry=Entry(SDEL,bg="black",fg="#00FFFF",font=("Calibri", 20), borderwidth=1,textvariable=ENRDEL)
        #DELENR_entry.place(x=170,y=100)
        global LAB1
        def LI(e):
            sdt=PhotoImage(file="images/buttons/delete2.png")
            Dbutton.image=sdt
            Dbutton.config(image=sdt,borderwidth=0)
        def li(e):
            sdt=PhotoImage(file="images/buttons/delete1.png")
            Dbutton.image=sdt
            Dbutton.config(image=sdt,borderwidth=0)

        def IR(e):
            sdt1=PhotoImage(file="images/buttons/c21.png")
            Cbutton.image=sdt1
            Cbutton.config(image=sdt1,borderwidth=0)
            
        def ir(e):
            sdt1=PhotoImage(file="images/buttons/c1.png")
            Cbutton.image=sdt1
            Cbutton.config(image=sdt1)
             
        def remove():
            os.remove('STUDENT_DETAILS/'+Yr+'/'+DELENR_info)
            list_of_files = os.listdir('dataset/'+Yr)
            for std in list_of_files :
                if DELENR_info ==(os.path.split(std)[1].split(".")[2]):
                    os.remove('dataset/'+Yr+'/'+std)
            LAB1=Label(tab2, text="",bg="black", fg="#00FFFF", font=("calibri", 20)).place(x=130,y=500)
            LAB1=Label(tab2, text="The Deletion Successful ",width=30,bg="black", fg="#00FFFF", font=("calibri", 20)).place(x=130,y=500)
            SDEL.destroy()
            Dbutton.place(x=1000,y=10000)
            Cbutton.place(x=1000,y=10300)
        def cancel():
            SDEL.destroy()
            Dbutton.place(x=1000,y=10000)
            Cbutton.place(x=1000,y=10300)
        
        global sdt
        global sdt1
        sdt1=PhotoImage(file="images/buttons/c1.png")
        Cbutton = tk.Button(SDEL, image=sdt1,command = cancel, borderwidth=0,bg="black",activebackground='black')
        Cbutton.bind("<Enter>", IR)
        Cbutton.bind("<Leave>", ir)
        Cbutton.place(x=100,y=230)

        sdt=PhotoImage(file="images/buttons/delete1.png")
        Dbutton = tk.Button(SDEL, image=sdt,bg="black" ,command=remove,activebackground='black',borderwidth=0)
        Dbutton.bind("<Enter>", LI)
        Dbutton.bind("<Leave>", li)
        Dbutton.place(x=300,y=230)
        
    else:
        LAB1=Label(tab2, text="",bg="black", fg="#00FFFF", font=("RED", 20)).place(x=130,y=500)
        LAB1=Label(tab2, text="The Details are not found ",width=30,bg="black", fg="RED", font=("calibri", 20)).place(x=130,y=500)

def Show_det():

    SDEL=Frame(tab2,bg="black")
    SDEL.place(x=800,y=80,height=500,width=500)

    tk.Label(SDEL, text="PLEASE ENTER STUDENT DETAILS  ",fg="#00ffcc",bg="black",font=("Calibri", 20)).place(x=70,y=5)
    global ENRDEL
    ENRDEL = StringVar()

    Label(SDEL, text="ENR NO :",bg="black",fg="white",font=("Calibri", 18), borderwidth=0).place(x=40,y=100)

    DELENR_entry=Entry(SDEL,bg="black",fg="#00FFFF",font=("Calibri", 20), borderwidth=1,textvariable=ENRDEL)
    DELENR_entry.place(x=170,y=100)

    def LI(e):
        sdt=PhotoImage(file="images/buttons/delete2.png")
        Dbutton.image=sdt
        Dbutton.config(image=sdt,borderwidth=0)
    def li(e):
        sdt=PhotoImage(file="images/buttons/delete1.png")
        Dbutton.image=sdt
        Dbutton.config(image=sdt,borderwidth=0)

    global sdt
    sdt=PhotoImage(file="images/buttons/delete1.png")
    Dbutton = tk.Button(tab2, image=sdt,bg="black" ,activebackground='black',borderwidth=0)
    Dbutton.bind("<Enter>", LI)
    Dbutton.bind("<Leave>", li)
    Dbutton.place(x=900,y=600)

####################### tab3 functions: Functions for subject reg/del #######################################################################################

def tab3clr():
    Subname_entry.delete(0, END)
    CC_entry.delete(0, END)
    Sem_entry.delete(0, END)
    Lecs_entry.delete(0, END)

def register_subjects():
        
    Subname_info = Subname.get()
    CC_info = CC.get()
    Sem_info = Sem.get()
    Lecs_info = Lecs.get()
    Year_info=tab3selection
    
    if Year_info == '1':
        Yr = 'FY'
        #print('assigned 1')

    elif Year_info == '2':
        Yr = 'SY'
        #print('assigned 2')

    elif Year_info == '3':
        Yr = 'TY'
        #print('assigned 3')

    if not os.path.exists('Teachers/'+userName+'/COURSES/'+Yr+'/'):
        os.makedirs('Teachers/'+userName+'/COURSES/'+Yr+'/')
    
    if os.path.exists('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info):
        global lb
        lb=Label(tab3, text="",bg="black", fg="#00FFFF", font=("calibri", 20)).place(x=550,y=620)
        lb=Label(tab3, text="The Subject already registered",bg="black", fg="red", font=("calibri", 20)).place(x=550,y=620)
        Subname_entry.delete(0, END)
        CC_entry.delete(0, END)
        Sem_entry.delete(0, END)
        Lecs_entry.delete(0, END)
    
    if not os.path.exists('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info):
        os.makedirs('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info)
        file = open('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info+'/'+CC_info, "w")
        file.write(Subname_info + "\n")
        file.write(CC_info + "\n")
        file.write(Sem_info + "\n")
        file.write(Lecs_info + "\n")
        file.close()
        lb=Label(tab3, text="",bg="black", fg="#00FFFF", font=("calibri", 20)).place(x=550,y=620)
        lb=Label(tab3, text="Course Registration Success",bg="black", fg="#00FFFF", font=("calibri", 20)).place(x=550,y=620)
        Subname_entry.delete(0, END)
        CC_entry.delete(0, END)
        Sem_entry.delete(0, END)
        Lecs_entry.delete(0, END)

############## tab4 functions #########################################################################################################################################

def tab4clr():
    DELCC_entry.delete(0, END)

def Del_course():
    global LAB2
    DELCC_info = CCDEL.get()

    Year_info=tab4selection
    
    global Yr
    if Year_info == '1':
        Yr = 'FY'
        #print('assigned 1')

    elif Year_info == '2':
        Yr = 'SY'
        #print('assigned 2')

    elif Year_info == '3':
        Yr = 'TY'
        #print('assigned 3')

    if os.path.exists('Teachers/'+userName+'/COURSES/'+Yr+'/'+DELCC_info):
        LAB2=Label(tab4, text="",bg="black", fg="#00FFFF", font=("calibri", 20))
        LAB2.place(x=130,y=500)
        LAB2.configure(text="Details are Founded",width=30)
        SUBDEL=Frame(tab4,bg="black",highlightbackground="orange",highlightcolor='#24f2e8', highlightthickness=2)
        SUBDEL.place(x=800,y=80,height=500,width=500)
        tk.Label(SUBDEL, text="THE FOUNDED DETAILS ",fg="#00ffcc",bg="black",font=("Calibri", 20)).place(x=120,y=5)
        #global CCDEL
        #CCDEL = StringVar()
        global det
        det=[]
        with open('Teachers/'+userName+'/COURSES/'+Yr+'/'+DELCC_info+'/'+DELCC_info, 'r') as filehandle:
            for line in filehandle:
                det.append(line)
                ##print(line)
        ##print(det)
        newdet = [x[:-1] for x in det]
        ##print(newdet)
        Label(SUBDEL, text="COURSE NAME :",bg="black",fg="white",font=("Calibri", 18), borderwidth=0).place(x=40,y=100)
        Label(SUBDEL, text=newdet[0],bg="black",fg="white",font=("Calibri", 18), borderwidth=0).place(x=250,y=100)
        Label(SUBDEL, text="COURSE CODE :",bg="black",fg="white",font=("Calibri", 18), borderwidth=0).place(x=40,y=180)
        Label(SUBDEL, text=newdet[1],bg="black",fg="white",font=("Calibri", 18), borderwidth=0).place(x=250,y=180)
        Label(SUBDEL, text="semester:",bg="black",fg="white",font=("Calibri", 18), borderwidth=0).place(x=40,y=260)
        Label(SUBDEL, text=newdet[2],bg="black",fg="white",font=("Calibri", 18), borderwidth=0).place(x=250,y=260)
        Label(SUBDEL, text="CREDITS :",bg="black",fg="white",font=("Calibri", 18), borderwidth=0).place(x=40,y=340)
        Label(SUBDEL, text=newdet[3],bg="black",fg="white",font=("Calibri", 18), borderwidth=0).place(x=250,y=340)

        def LI(e):
            subdt=PhotoImage(file="images/buttons/delete2.png")
            Dbutton.image=subdt
            Dbutton.config(image=subdt,borderwidth=0)
        def li(e):
            subdt=PhotoImage(file="images/buttons/delete1.png")
            Dbutton.image=subdt
            Dbutton.config(image=subdt,borderwidth=0)

        def IR(e):
            subd1=PhotoImage(file="images/buttons/clear2.png")
            Cbutton.image=subd1
            Cbutton.config(image=subd1,borderwidth=0)
            
        def ir(e):
            subd1=PhotoImage(file="images/buttons/clear1.png")
            Cbutton.image=subd1
            Cbutton.config(image=subd1)
            
        def remove():
            shutil.rmtree('Teachers/'+userName+'/COURSES/'+Yr+'/'+DELCC_info)
            LAB2=Label(tab4, text="",bg="black", fg="#00FFFF", font=("calibri", 20)).place(x=130,y=500)
            LAB2=Label(tab4, text="The Deletion Successful ",width=30,bg="black", fg="#00FFFF", font=("calibri", 20))
            LAB2.place(x=130,y=500)
            SUBDEL.destroy()
            Dbutton.place(x=1000,y=10000)
            Cbutton.place(x=1000,y=10300)
            
        def cancel():
            SUBDEL.destroy()
            Dbutton.place(x=1000,y=10000)
            Cbutton.place(x=1000,y=10300)
            
        global subd1
        subd1=PhotoImage(file="images/buttons/clear1.png")
        Cbutton = tk.Button(tab4, image=subd1,command = cancel, borderwidth=0,bg="black",activebackground='black')
        Cbutton.bind("<Enter>", IR)
        Cbutton.bind("<Leave>", ir)
        Cbutton.place(x=840,y=550)

        global subdt
        subdt=PhotoImage(file="images/buttons/delete1.png")
        Dbutton = tk.Button(tab4, image=subdt,bg="black" ,command=remove,activebackground='black',borderwidth=0)
        Dbutton.bind("<Enter>", LI)
        Dbutton.bind("<Leave>", li)
        Dbutton.place(x=1070,y=550)
 
        #os.remove('STUDENT_DETAILS/'+DELCC_info)
        #list_of_files = os.listdir('dataset')
        #for std in list_of_files :
            #if DELCC_info ==(os.path.split(std)[1].split(".")[2]):
                #os.remove('dataset/'+std)
        #LAB2=Label(tab4, text="",bg="black", fg="#00FFFF", font=("calibri", 20)).place(x=200,y=320)
        #LAB2=Label(tab4, text="The Deletion Successful ",bg="black", fg="#00FFFF", font=("calibri", 20)).place(x=200,y=320)
        
    else:
        
        LAB2=Label(tab4, text="",bg="black", fg="#00FFFF", font=("RED", 20))
        LAB2.place(x=130,y=500)
        LAB2=Label(tab4, text="The Details are not found ",width=30,bg="black", fg="RED", font=("calibri", 20))
        LAB2.place(x=130,y=500)
        #tab4.after(1000)
        #LAB2.configure(text="")
        
######################################################################################################################################################
####################################### home tab #####################################################################################################

fname = PhotoImage(file = "images/bg/ho.png")
background_label = Label(Home, image=fname)
background_label.place(x=0, y=0, relwidth=1, relheight=1)

#user data frame
USR=Frame(Home,bg='black')
USR.place(x=0,y=0,height=900,width=300)

HRL=Frame(USR,bg='grey')
HRL.place(x=0,y=250,height=2,width=300)

HRL2=Frame(USR,bg='grey')
HRL2.place(x=0,y=450,height=2,width=300)

HRBT=Frame(Home,bg='grey')
HRBT.place(x=0,y=0,height=2,width=1400)

HRBF=Frame(Home,bg='grey')
HRBF.place(x=0,y=660,height=2,width=1400)

#VR1=Frame(Home,bg='white')
#VR1.place(x=0,y=0,height=900,width=2)

VRL=Frame(Home,bg='grey')
VRL.place(x=1360,y=0,height=900,width=2)

VR=Frame(Home,bg='grey')
VR.place(x=300,y=0,height=900,width=2)

VRM=Frame(Home,bg='grey')
VRM.place(x=820,y=0,height=900,width=2)

MHR=Frame(Home,bg='grey')
MHR.place(x=300,y=330,height=2,width=1400)

HR=Frame(USR,bg='#0e1111',highlightbackground="white", highlightthickness=2)
HR.place(x=0,y=0,height=250,width=300)

HRB=Frame(USR,bg='#0e1111',highlightbackground="white", highlightthickness=2)
HRB.place(x=0,y=450,height=250,width=300)

def UploadActionh():
    global Filenameh
    Filenameh = filedialog.askopenfilename(title="Choose a file",filetypes=[('image files', ('.png', '.jpg'))])
    if (pathlib.Path(Filenameh).suffix) != '.png':
        im1h = Image.open(r''+Filenameh)
        imh = im1h.resize((100, 111))
        imh.save(r'Teachers/'+userName+'/user.png')

    else:
        im1h = Image.open(r''+Filenameh)
        imh = im1h.resize((100, 111))
        imh.save(r'Teachers/'+userName+'/user.png')
        #shutil.copy(Filenameh,'Teachers/'+userName+'/user.png')
    profilePh()

def profilePh():
    global Ppimg
    if os.path.exists('Teachers/'+userName+'/user.png'):
        Ppimg=PhotoImage(file='Teachers/'+userName+'/user.png')
        buttonp = tk.Button(USR, text='Open', command=UploadActionh, image=Ppimg,bg="#0e1111", borderwidth=0).place(x=100,y=50)
    else:
        Ppimg=PhotoImage(file="images/icons/user.png")
        buttonp = tk.Button(USR, text='Open', command=UploadActionh, image=Ppimg,bg="#0e1111", borderwidth=0).place(x=100,y=50)

profilePh()

#Homeimagei=PhotoImage(file="images/icons/dp1.png")
#Homebuttoni = tk.Button(USR, image=Homeimagei,bg="#0e1111",activebackground='white', borderwidth=0)
#Homebuttoni.place(x=70,y=30)

Label(USR, text="ABOUT US ",bg="black",fg="orange",font=("Calibri", 14,'bold'), borderwidth=0).place(x=100,y=260)

Label(USR, text="Jainam Patel",bg="black",fg="white",font=("Calibri", 14,'bold'), borderwidth=0).place(x=30,y=350)
Label(USR, text="2062077",bg="black",fg="white",font=("Calibri", 14,'bold'), borderwidth=0).place(x=150,y=350)

Label(HRB, text="CONTACT US ",bg="#0e1111",fg="orange",font=("Calibri", 14,'bold'), borderwidth=0).place(x=90,y=10)

import webbrowser

new = 1
url = "https://chat.whatsapp.com/IqfUUbRcrooHTcHXt9q2uR"

def openweb():
    webbrowser.open(url,new=new)

new1 = 1
url1 = "https://m.youtube.com/channel/UCM3I84mdVINaKoXELyrjJKQ"

def openweb1():
    webbrowser.open(url1,new=new1)

Label(HRB, text="For Any Queries conact us",bg="#0e1111",fg="white",font=("Calibri", 14,'bold'), borderwidth=0).place(x=50,y=50)

whatsapp=PhotoImage(file="images/icons/w.png")
whatsappi = tk.Button(HRB, image=whatsapp,bg="#0e1111",activebackground='#0e1111', borderwidth=0,command=openweb)
whatsappi.place(x=20,y=100)

insta=PhotoImage(file="images/icons/insta.png")
instai = tk.Button(HRB, image=insta,bg="#0e1111",activebackground='#0e1111', borderwidth=0)
instai.place(x=90,y=100)

facebook=PhotoImage(file="images/icons/fb.png")
facebooki = tk.Button(HRB, image=facebook,bg="#0e1111",activebackground='#0e1111', borderwidth=0)
facebooki.place(x=160,y=100)

youtubei=PhotoImage(file="images/icons/youtube.png")
youtube = tk.Button(HRB, image=youtubei,bg="#0e1111",activebackground='#0e1111', borderwidth=0,command=openweb1)
youtube.place(x=230,y=100)

if os.path.exists('Teachers/'+userName+'/'+userName):
    global usrdet
    usrdet=[]
    with open('Teachers/'+userName+'/'+userName, 'r') as usrfile:
        for line in usrfile:
            usrdet.append(line)
    newusrdet = [x[:-1] for x in usrdet]
    #Label(USR, text="NAME :",bg="white",fg="white",font=("Calibri", 14), borderwidth=0).place(x=20,y=200)
    Label(USR, text=newusrdet[0],bg="#0e1111",fg="white",font=("Calibri", 14,'bold'), borderwidth=0).place(x=120,y=180)
    #Label(USR, text="Email id:",bg="white",fg="white",font=("Calibri", 14), borderwidth=0).place(x=20,y=230)
    Label(USR, text=newusrdet[1],bg="#0e1111",fg="white",font=("Calibri", 14,'bold'), borderwidth=0).place(x=50,y=210)

if os.path.exists('Teachers/'+userName+'/COURSES'):
    global cdet,crsdet
    cdet=[]
    crsdet=[]
    Label(USR, text="ABOUT US ",bg="black",fg="orange",font=("Calibri", 14,'bold'), borderwidth=0).place(x=100,y=260)
    cdet= os.listdir('Teachers/'+userName+'/COURSES')
    #print(cdet)
    #Label(USR, text="Courses",bg="black",fg="red",font=("Calibri", 12,'bold'), borderwidth=0).place(x=10,y=300)
    #Label(USR, text="Course-code",bg="black",fg="red",font=("Calibri", 12,'bold'), borderwidth=0).place(x=120,y=300)
    #Label(USR, text="Credits",bg="black",fg="red",font=("Calibri", 12,'bold'), borderwidth=0).place(x=230,y=300)
    
    yax=330
    for c in cdet:
        cos= os.listdir('Teachers/'+userName+'/COURSES/'+c)
        for cd in cos:
            with open('Teachers/'+userName+'/COURSES/'+c+'/'+cd+'/'+cd, 'r') as crsfile:
                for line in crsfile:
                    crsdet.append(line)
            newcrsdet = [x[:-1] for x in crsdet]
            #print(newcrsdet[0])
            #Label(USR, text=newcrsdet[0],bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=10,y=yax)
            #Label(USR, text=newcrsdet[1],bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=120,y=yax)
            #Label(USR, text=newcrsdet[3],bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=260,y=yax)
            #newcrsdet.clear()
            #crsdet.clear()
            #yax=yax+30

#user data frame end

#charts start
        
from pandas import DataFrame
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

Month=['Start','Jan','Feb','Mar','APR','MAY']
Average_Attendance_Rate=[0,9,8,7,6,4]

#per course the student present in the lecture 
CAVG=Frame(Home,bg="black",highlightbackground="orange",highlightcolor='#24f2e8', highlightthickness=2)
CAVG.place(x=350,y=10,height=300,width=400)

#how much portion is completed
COMPL=Frame(Home,bg="black",highlightbackground="orange",highlightcolor='#24f2e8', highlightthickness=2)
COMPL.place(x=900,y=350,height=300,width=400)

AVG=Frame(Home,bg="black",highlightbackground="orange",highlightcolor='#24f2e8', highlightthickness=2)
AVG.place(x=900,y=10,height=300,width=400)

cd=[]
av=[]
if os.path.exists('Teachers/'+userName+'/COURSES'):
    global cavg,crsavg
    cavg=[]
    crsavg=[]
    cavg= os.listdir('Teachers/'+userName+'/COURSES')
    #print(cavg)
    for i in cavg:
        for j in cavg:
            coavg=os.listdir('Teachers/'+userName+'/COURSES/'+j)
            for crsav in coavg :
                if os.path.exists('Teachers/'+userName+'/COURSES/'+j+'/'+crsav+'/Regular_Attendance/'+crsav+'_Regular_Attendance.csv'):
                    cf = pd.read_csv('Teachers/'+userName+'/COURSES/'+j+'/'+crsav+'/Regular_Attendance/'+crsav+'_Regular_Attendance.csv')
                    matrix2 = cf[cf.columns[7]].to_numpy()
                    list2 = matrix2.tolist()
                    ##print(list2)
                    cd.append(crsav)
                    av.append(sum(list2)/len(list2))
                else:
                    cd.append(crsav)
                    av.append(0)

        #print(cd)
        #print(av)
        avgdict={}
        for a in range(0,len(cd)):
            avgdict[cd[a]]=av[a]
            #print(avgdict)
        Label(AVG, text="THE COURSES DETAILS ",bg="black",fg="orange",font=("Calibri", 16), borderwidth=0).place(x=100,y=10)
        Label(AVG, text="Courses",bg="black",fg="red",font=("Calibri", 14,'bold'), borderwidth=0).place(x=10,y=60)
        Label(AVG, text="C_Code",bg="black",fg="red",font=("Calibri", 14,'bold'), borderwidth=0).place(x=120,y=60)
        Label(AVG, text="Credits",bg="black",fg="red",font=("Calibri", 14,'bold'), borderwidth=0).place(x=210,y=60)
        Label(AVG, text="Avgerage ",bg="black",fg="red",font=("Calibri", 14,'bold'), borderwidth=0).place(x=300,y=60)
        #Label(AVG, text="Rem_lec",bg="black",fg="red",font=("Calibri", 14,'bold'), borderwidth=0).place(x=380,y=60)
        cdet=[]
        crsdet=[]
        cdet= os.listdir('Teachers/'+userName+'/COURSES')
        yax=100
        for cdf in cdet:
            yc= os.listdir('Teachers/'+userName+'/COURSES/'+cdf)
            for c in yc:
                with open('Teachers/'+userName+'/COURSES/'+cdf+'/'+c+'/'+c, 'r') as crsfile:
                    for line in crsfile:
                        crsdet.append(line)
                        newcrsdet = [x[:-1] for x in crsdet]
                    #print(newcrsdet)
                    Label(AVG, text=newcrsdet[0],bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=10,y=yax)
                    Label(AVG, text=newcrsdet[1],bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=120,y=yax)
                    Label(AVG, text=newcrsdet[3],bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=220,y=yax)
                    Label(AVG, text=avgdict[newcrsdet[1]],bg="black",fg="white",font=("Calibri", 12), borderwidth=0).place(x=300,y=yax)
                    yax=yax+30

                newcrsdet.clear()
                crsdet.clear()
                
##print(cd)
##print(av)

#Label(Home, text="THE AVERAGE ATTENDANCE PER COURSE",bg="black",fg="orange",font=("Calibri", 16), borderwidth=0).place(x=380,y=20)            

#first chart
with plt.style.context('dark_background'):
    figure1 = plt.Figure(figsize=(4,3), dpi=100)
    ax1 = figure1.add_subplot(111)
    ax1.set_facecolor("black")

bar1 = FigureCanvasTkAgg(figure1, CAVG)
bar1.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
ax1.bar(cd,av,color='#24f2e8',width=0.3)
ax1.set_title('Average_Attendance Per Course',color='orange')
ax1.legend(['Average_Attendance'])
ax1.set_xlabel('Courses')
figure1.autofmt_xdate(rotation=30)
#1st chart end 

#second chart
with plt.style.context('dark_background'):
    figure2 = plt.Figure(figsize=(4,3), dpi=100)
    ax2 = figure2.add_subplot(111)
    ax2.set_facecolor("black")

line2 = FigureCanvasTkAgg(figure2, COMPL)
line2.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
#df2 = df2(Month,Average_Attendance_Rate)
#df2.plot(kind='line', legend=True, ax=ax2, color='r',marker='o', fontsize=10)
ax2.plot(cd,av,color='#24f2e8', marker='o')
ax2.set_title('Average_Attendance_Rate per course ',color='orange')
ax2.legend(['Average_Attendance_Rate'])
ax2.set_xlabel('Course')
figure2.autofmt_xdate(rotation=0)
#2nd chart end 

CAVGPIE=Frame(Home,bg="black",highlightbackground="orange",highlightcolor='#24f2e8', highlightthickness=2)
CAVGPIE.place(x=350,y=350,height=300,width=400)
    #Label(CAVGPIE, text="THE homeavERAGE ATTENDANCE of COURSES",bg="black",fg="aqua",font=("Calibri", 16), borderwidth=0).place(x=0,y=0)            

    #3rd chart
with plt.style.context('dark_background'):
    homefigure2 = plt.Figure(figsize=(4,3), dpi=100)
    homeax2 = homefigure2.add_subplot(111)
    homeax2.set_facecolor("white")

homebar1 = FigureCanvasTkAgg(homefigure2, CAVGPIE)
homebar1.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
explode = (0, 0.1)
if len(av)!=0 :
    homeax2.pie([sum(av)/len(av),100-sum(av)/len(av)],explode=explode,labels=["Present","Absent"],colors=['Orange','white'],autopct='%1.1f%%',shadow=True, startangle=90)
    homeax2.set_title('Average_Attendance_Of All Courses',color='orange')
    homeax2.legend(["Present : "+str(sum(av)/len(av)),"Absent : "+str(100-sum(av)/len(av))])
    #homeax2.set_xlabel('Courses')
    #3rd chart end 
                   
######## Home tab end #####################################################################################################################################

############### tab1 start #####################################################################################################

ftab1=PhotoImage(file = "images/bg/ho.png")
background_label = Label(tab1, image=ftab1)
background_label.place(x=0, y=0, relwidth=1, relheight=1)

global Sname
global ENR
global Email
global Phone
global SM
global YEAR

global Sname_entry
global ENR_entry
global Email_entry
global Phone_entry
global SEM_entry

#R1=Frame(tab1,bg='#24f2e8')
#R1.place(x=0,y=0,height=900,width=300)

Sname = StringVar()
ENR = StringVar()
Email = StringVar()
Phone = StringVar()
PEmail = StringVar()
PPhone = StringVar()

SM = StringVar()
SREG=Frame(tab1,bg="black",highlightbackground="orange",highlightcolor='#24f2e8', highlightthickness=2)
SREG.place(x=250,y=50,height=550,width=900)
tk.Label(SREG, text="PLEASE ENTER STUDENT DEATAILS FOR REGISTRATION", height="2",fg="#24f2e8",bg="black",font=("Calibri", 22)).place(x=180,y=10)
Label(SREG, text="STUDENT NAME :",fg="white",bg="black",font=("Calibri", 16), borderwidth=0).place(x=250,y=100)
Label(SREG, text="ENR NO:",fg="white",bg="black",font=("Calibri", 16), borderwidth=0).place(x=250,y=160)

HRLs=Frame(SREG,bg='orange')
HRLs.place(x=0,y=200,height=2,width=900)

Label(SREG, text="EMAIL ID :",fg="white",bg="black",font=("Calibri", 16), borderwidth=0).place(x=50,y=220)
Label(SREG, text="PHONE NO:",fg="white",bg="black",font=("Calibri", 16), borderwidth=0).place(x=50,y=280)

Label(SREG, text="PARENT EMAIL :",fg="white",bg="black",font=("Calibri", 16), borderwidth=0).place(x=450,y=220)
Label(SREG, text="PARENT PHONE :",fg="white",bg="black",font=("Calibri", 16), borderwidth=0).place(x=450,y=280)

HRLs2=Frame(SREG,bg='orange')
HRLs2.place(x=0,y=340,height=2,width=900)

Label(SREG, text="YEAR :",fg="white",bg="black",font=("Calibri", 16), borderwidth=0).place(x=250,y=360)
Label(SREG, text="SEM :",fg="white",bg="black",font=("Calibri", 16), borderwidth=0).place(x=250,y=420)

Sname_entry=Entry(SREG,font=("Calibri", 16),fg="#00FFFF",bg="black",textvariable=Sname, borderwidth=1,insertbackground="#00FFFF")
Sname_entry.place(x=420,y=100)
ENR_entry=Entry(SREG,font=("Calibri", 16),fg="#00FFFF",bg="black",textvariable=ENR, borderwidth=1,insertbackground="#00FFFF")
ENR_entry.place(x=420,y=160)

Email_entry=Entry(SREG,font=("Calibri", 16),fg="#00FFFF",bg="black",textvariable=Email, borderwidth=1,insertbackground="#00FFFF")
Email_entry.place(x=160,y=220)
Phone_entry=Entry(SREG,font=("Calibri", 16),fg="#00FFFF",bg="black",textvariable=Phone, borderwidth=1,insertbackground="#00FFFF")
Phone_entry.place(x=160,y=280)

PEmail_entry=Entry(SREG,font=("Calibri", 16),fg="#00FFFF",bg="black",textvariable=PEmail, borderwidth=1,insertbackground="#00FFFF")
PEmail_entry.place(x=600,y=220)
PPhone_entry=Entry(SREG,font=("Calibri", 16),fg="#00FFFF",bg="black",textvariable=PPhone, borderwidth=1,insertbackground="#00FFFF")
PPhone_entry.place(x=600,y=280)

vari = StringVar()
vari.set(1)

def quit_loop():
    #print ("Selection:"+vari.get())
    global selection
    selection = vari.get()

Radiobutton(SREG,text="1st year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=vari, value = 1).place(x=420,y=360)
Radiobutton(SREG,  text="2nd year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=vari, value = 2).place(x=530,y=360)
Radiobutton(SREG,  text="3rd year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=vari, value = 3).place(x=630,y=360)

try:
    selection
except NameError:
    #print("well, it WASN'T defined after all!")
    selection=1

SEM_entry=Entry(SREG,font=("Calibri", 16),fg="#00FFFF",bg="black",textvariable=SM, borderwidth=1,insertbackground="#00FFFF")
SEM_entry.place(x=420,y=420)

def IR(e):
    t1imageR=PhotoImage(file="images/buttons/clear2.png")
    t1buttonR.config(image=t1imageR,borderwidth=0)
    t1buttonR.image=t1imageR
def ir(e):
    t1imageR=PhotoImage(file="images/buttons/clear1.png")
    t1buttonR.config(image=t1imageR)
    t1buttonR.image=t1imageR
t1imageR=PhotoImage(file="images/buttons/clear1.png")
t1buttonR = tk.Button(tab1, image=t1imageR,command = tab1clr, borderwidth=0,bg="black",activebackground='black')
t1buttonR.bind("<Enter>", IR)
t1buttonR.bind("<Leave>", ir)

def IL(e):
    t1imageL=PhotoImage(file="images/buttons/submit2.png")
    t1buttonL.config(image=t1imageL)
    t1buttonL.image=t1imageL
def il(e):
    t1imageL=PhotoImage(file="images/buttons/submit1.png")
    t1buttonL.config(image=t1imageL)
    t1buttonL.image=t1imageL
t1imageL=PhotoImage(file="images/buttons/submit1.png")
t1buttonL = tk.Button(tab1, image=t1imageL, borderwidth=0,command=lambda:[quit_loop(),register_student()],bg="black",activebackground='black')
t1buttonL.bind("<Enter>", IL)
t1buttonL.bind("<Leave>", il)

t1buttonR.place(x=400,y=540)
t1buttonL.place(x=800,y=540)

t1imagei=PhotoImage(file="images/icons/student.png")
t1buttoni = tk.Button(tab1, image=t1imagei,bg="black",activebackground='black', borderwidth=0)
t1buttoni.place(x=300,y=100)

############### tab1 end #####################################################################################################

######### tab2 start ################################################################################################################################################################

def tab2stdavg():
    #print("from tab2stdavg ")
    if os.path.exists('Teachers/'+userName+'/COURSES'):
        DELENR_info = ENRDEL.get()
        Year_info=tab2selection
        if Year_info == '1':
            Yr = 'FY'
            ##print('assigned 1')

        elif Year_info == '2':
            Yr = 'SY'
            ##print('assigned 2')

        elif Year_info == '3':
            Yr = 'TY'
            ##print('assigned 3')
        
        global savg
        savg=[]
        c=[]
        a=[]
        savg = os.listdir('Teachers/'+userName+'/COURSES')
        for i in savg:
            courses=os.listdir('Teachers/'+userName+'/COURSES/'+i)
            for j in courses:
                if os.path.exists('Teachers/'+userName+'/COURSES/'+i+'/'+j+'/Regular_Attendance/'+j+'_Regular_Attendance.csv'):
                    cf = pd.read_csv('Teachers/'+userName+'/COURSES/'+i+'/'+j+'/Regular_Attendance/'+j+'_Regular_Attendance.csv') 
                    matrix1 = cf[cf.columns[0]].to_numpy()
                    list1 = matrix1.tolist()
                    #print("list 1")
                    #print(list1)                    
                    matrix2 = cf[cf.columns[7]].to_numpy()
                    list2 = matrix2.tolist()
                    #print("list 2")
                    #print(list2)
                    df=cf.set_index('ENR')
                    print(df)
                    if DELENR_info in list1:
                        print(df.at[str(DELENR_info), 'Attendance percentage'])
                        c.append(j)
                        a.append(df.at[str(DELENR_info), 'Attendance percentage'])

                    print(c)
                    print(a)
                    
    SAVG=Frame(tab2,bg="black",highlightbackground="orange",highlightcolor='#24f2e8', highlightthickness=2)
    SAVG.place(x=120,y=350,height=300,width=400)
    #Label(SAVG, text="THE tab2avERAGE ATTENDANCE PER COURSE",bg="black",fg="aqua",font=("Calibri", 16), borderwidth=0).place(x=0,y=0)            

    #first chart
    with plt.style.context('dark_background'):
        tab2figure1 = plt.Figure(figsize=(4,3), dpi=100)
        tab2ax1 = tab2figure1.add_subplot(111)
        tab2ax1.set_facecolor("white")

    tab2bar1 = FigureCanvasTkAgg(tab2figure1, SAVG)
    tab2bar1.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    if os.path.exists('Teachers/'+userName+'/COURSES'):
        tab2ax1.bar(c,a,color='#00FF00',width=0.3)
    #tab2ax1.set_title('Courses Vs. tab2average_Attendance')
    tab2ax1.legend(['tab2average_Attendance'])
    tab2ax1.set_xlabel('Courses')
    tab2figure1.autofmt_xdate(rotation=0)
    #1st chart end

    CAVG=Frame(tab2,bg="black",highlightbackground="orange",highlightcolor='#24f2e8', highlightthickness=2)
    CAVG.place(x=850,y=350,height=300,width=400)
    #Label(CAVG, text="THE tab2avERAGE ATTENDANCE of COURSES",bg="black",fg="aqua",font=("Calibri", 16), borderwidth=0).place(x=0,y=0)            

    #2nd chart
    with plt.style.context('dark_background'):
        tab2figure2 = plt.Figure(figsize=(4,3), dpi=100)
        tab2ax2 = tab2figure2.add_subplot(111)
        tab2ax2.set_facecolor("white")

    tab2bar1 = FigureCanvasTkAgg(tab2figure2, CAVG)
    tab2bar1.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    if os.path.exists('Teachers/'+userName+'/COURSES'):
        if len(a) != 0:
            tab2ax2.pie([sum(a)/len(a),100-sum(a)/len(a)],explode=explode,labels=["Present","Absent"],colors=['Orange','white'],autopct='%1.1f%%',shadow=True, startangle=90)
            tab2ax2.legend(["Present : "+str(sum(a)/len(a)),"Absent : "+str(100-sum(a)/len(a))])
    #tab2ax2.set_xlabel('Courses')
    #2nd chart end 
                   
ftab2=PhotoImage(file = "images/bg/ho.png")
background_label = Label(tab2, image=ftab2)
background_label.place(x=0, y=0, relwidth=1, relheight=1)

tab2VL=Frame(tab2,bg='white')
tab2VL.place(x=700,y=0,height=900,width=2)

tab2MHR=Frame(tab2,bg='white')
tab2MHR.place(x=0,y=340,height=2,width=1400)

SDEL=Frame(tab2,bg="black",highlightbackground="orange", highlightcolor="orange", highlightthickness=2)
SDEL.place(x=100,y=10,height=300,width=500)

#SD=Frame(tab2,bg="#24f2e8")
#SD.place(x=0,y=10,height=300,width=150)

tk.Label(SDEL, text="PLEASE ENTER STUDENT ENR NO.  ",fg="#24f2e8",bg="black",font=("Calibri", 18)).place(x=70,y=20)
global ENRDEL
ENRDEL = StringVar()

Label(SDEL, text="Year:",bg="black",fg="white",font=("Calibri", 16), borderwidth=0).place(x=50,y=90)

tab2year = StringVar()
tab2year.set(1)

def quit_loop_tab2():
    #print ("tab2selection:"+tab2year.get())
    global tab2selection
    tab2selection = tab2year.get()

Radiobutton(SDEL,text="1st year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=tab2year, value = 1).place(x=150,y=90)
Radiobutton(SDEL,  text="2nd year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=tab2year, value = 2).place(x=250,y=90)
Radiobutton(SDEL,  text="3rd year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=tab2year, value = 3).place(x=350,y=90)

try:
    tab2selection
except NameError:
    #print("well, it WASN'T defined after all!")
    tab2selection=1
    
Label(SDEL, text="ENR NO ",bg="black",fg="white",font=("Calibri", 16), borderwidth=0).place(x=50,y=170)

DELENR_entry=Entry(SDEL,bg="black",fg="#00FFFF",font=("Calibri", 16), borderwidth=1,textvariable=ENRDEL,insertbackground="#00FFFF")
DELENR_entry.place(x=150,y=170)

def LI(e):
    t2imageL=PhotoImage(file="images/buttons/clear2.png")
    t2buttonL.image=t2imageL
    t2buttonL.config(image=t2imageL,borderwidth=0)
def li(e):
    t2imageL=PhotoImage(file="images/buttons/clear1.png")
    t2buttonL.image=t2imageL
    t2buttonL.config(image=t2imageL,borderwidth=0)
t2imageL=PhotoImage(file="images/buttons/clear1.png")
t2buttonL = tk.Button(tab2, image=t2imageL,bg="black",command=tab2clr ,activebackground='black',borderwidth=0)
t2buttonL.bind("<Enter>", LI)
t2buttonL.bind("<Leave>", li)

def RI(e):
    t2imageR=PhotoImage(file="images/buttons/search2.png")
    t2buttonR.config(image=t2imageR)
    t2buttonR.image=t2imageR
def ri(e): 
    t2imageR=PhotoImage(file="images/buttons/search1.png")
    t2buttonR.config(image=t2imageR)
    t2buttonR.image=t2imageR
t2imageR=PhotoImage(file="images/buttons/search1.png")
t2buttonR = tk.Button(tab2, image=t2imageR,bg="black",command=lambda:[quit_loop_tab2(),del_student()],activebackground='black', borderwidth=0)
t2buttonR.bind("<Enter>", RI)
t2buttonR.bind("<Leave>", ri)

t2buttonL.place(x=150,y=250)
t2buttonR.place(x=400,y=250)

t2imagei=PhotoImage(file="images/icons/search.png")
t2buttoni = tk.Button(SDEL, image=t2imagei,bg="black",activebackground='black', borderwidth=0)
#t2buttoni.place(x=0,y=0)

##########tab2 end ###############################################################################################################################################################################################################################################################################################################################################################

######## tab3 starts ########################################################################################################################################################################################################################################

global Subname
global CC
global Sem
global Lecs

global Subname_entry
global CC_entry
global Sem_entry
global Lecs_entry

ftab3=PhotoImage(file = "images/bg/ho.png")
background_label = Label(tab3, image=ftab3)
background_label.place(x=0, y=0, relwidth=1, relheight=1)

Subname = StringVar()
CC = StringVar()
Sem = StringVar()
Lecs = StringVar()

SUBREG=Frame(tab3,bg="black",highlightbackground="orange", highlightcolor="orange", highlightthickness=2)
SUBREG.place(x=250,y=50,height=550,width=900)

tk.Label(SUBREG, text=" PLEASE ENTER DETAILS ABOUT THE COURSE YOU HAVE ", height="2",fg="#24f2e8",bg="black",font=("Calibri", 22)).place(x=180,y=10)
Label(SUBREG, text="Course Title :",fg="white",bg="black",font=("Calibri", 16), borderwidth=0).place(x=220,y=120)
Label(SUBREG, text="Course Code:",fg="white",bg="black",font=("Calibri", 16), borderwidth=0).place(x=220,y=200)
Label(SUBREG, text="Sem  :",fg="white",bg="black",font=("Calibri", 16), borderwidth=0).place(x=220,y=280)
Label(SUBREG, text="NO OF Credits:",fg="white",bg="black",font=("Calibri", 16), borderwidth=0).place(x=220,y=360)

Subname_entry=Entry(SUBREG,font=("Calibri", 16),fg="#00FFFF",bg="black",textvariable=Subname, borderwidth=1,insertbackground="#00FFFF")
Subname_entry.place(x=450,y=120)
CC_entry=Entry(SUBREG,font=("Calibri", 16),fg="#00FFFF",bg="black",textvariable=CC, borderwidth=1,insertbackground="#00FFFF")
CC_entry.place(x=450,y=200)
Sem_entry=Entry(SUBREG,font=("Calibri", 16),fg="#00FFFF",bg="black",textvariable=Sem, borderwidth=1,insertbackground="#00FFFF")
Sem_entry.place(x=450,y=280)
Lecs_entry=Entry(SUBREG,font=("Calibri", 16),fg="#00FFFF",bg="black",textvariable=Lecs, borderwidth=1,insertbackground="#00FFFF")
Lecs_entry.place(x=450,y=360)

Label(SUBREG, text="YEAR :",bg="black",fg="white",font=("Calibri", 16), borderwidth=0).place(x=220,y=450)

tab3year = StringVar()
tab3year.set(1)

def quit_loop_tab3():
    #print ("tab3selection:"+tab3year.get())
    global tab3selection
    tab3selection = tab3year.get()

Radiobutton(SUBREG,text="1st year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=tab3year, value = 1).place(x=450,y=450)
Radiobutton(SUBREG,  text="2nd year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=tab3year, value = 2).place(x=550,y=450)
Radiobutton(SUBREG,  text="3rd year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=tab3year, value = 3).place(x=650,y=450)

try:
    tab3selection
except NameError:
    #print("well, it WASN'T defined after all!")
    tab3selection=1

def IR(e):
    t3imageR=PhotoImage(file="images/buttons/clear2.png")
    t3buttonR.config(image=t3imageR,borderwidth=0)
    t3buttonR.image=t3imageR
def ir(e):
    t3imageR=PhotoImage(file="images/buttons/clear1.png")
    t3buttonR.config(image=t3imageR)
    t3buttonR.image=t3imageR
t3imageR=PhotoImage(file="images/buttons/clear1.png")
t3buttonR = tk.Button(tab3, image=t3imageR,command = tab3clr, borderwidth=0,bg="black",activebackground='black')
t3buttonR.bind("<Enter>", IR)
t3buttonR.bind("<Leave>", ir)

def IL(e):
    t3imageL=PhotoImage(file="images/buttons/submit2.png")
    t3buttonL.config(image=t3imageL)
    t3buttonL.image=t3imageL
def il(e):
    t3imageL=PhotoImage(file="images/buttons/submit1.png")
    t3buttonL.config(image=t3imageL)
    t3buttonL.image=t3imageL
t3imageL=PhotoImage(file="images/buttons/submit1.png")
t3buttonL = tk.Button(tab3, image=t3imageL, borderwidth=0,command=lambda:[quit_loop_tab3(),register_subjects()],bg="black",activebackground='black')
t3buttonL.bind("<Enter>", IL)
t3buttonL.bind("<Leave>", il)

t3buttonR.place(x=400,y=580)
t3buttonL.place(x=800,y=580)

t3imagei=PhotoImage(file="images/icons/course.png")
t3buttoni = tk.Button(tab3, image=t3imagei,bg="black",activebackground='black', borderwidth=0)
t3buttoni.place(x=300,y=100)

######### tab3 ends #######################################################################################################################

######### tab4 starts ################################################################################################################################################################

filename = PhotoImage(file = "images/bg/ho.png")
background_label = Label(tab4, image=filename)
background_label.place(x=0, y=0, relwidth=1, relheight=1)

tab4VL=Frame(tab4,bg='white')
tab4VL.place(x=700,y=0,height=900,width=2)

tab4MHR=Frame(tab4,bg='white')
tab4MHR.place(x=0,y=340,height=2,width=700)

SUBDEL=Frame(tab4,bg="black",highlightbackground="orange",highlightcolor='#24f2e8', highlightthickness=2)
SUBDEL.place(x=100,y=10,height=300,width=500)

tk.Label(SUBDEL, text="PLEASE ENTER COURSE CODE   ",fg="#24f2e8",bg="black",font=("Calibri", 18)).place(x=80,y=20)
global CCDEL
CCDEL = StringVar()

Label(SUBDEL, text="Year :",bg="black",fg="white",font=("Calibri", 16), borderwidth=0).place(x=50,y=90)

tab4year = StringVar()
tab4year.set(1)

def quit_loop_tab4():
    #print ("tab4selection:"+tab4year.get())
    global tab4selection
    tab4selection = tab4year.get()

Radiobutton(SUBDEL,text="1st year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=tab4year, value = 1).place(x=150,y=90)
Radiobutton(SUBDEL,  text="2nd year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=tab4year, value = 2).place(x=250,y=90)
Radiobutton(SUBDEL,  text="3rd year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=tab4year, value = 3).place(x=350,y=90)

try:
    tab4selection
except NameError:
    #print("well, it WASN'T defined after all!")
    tab4selection=1
    
Label(SUBDEL, text="COURSE CODE :",bg="black",fg="white",font=("Calibri", 16), borderwidth=0).place(x=50,y=170)

DELCC_entry=Entry(SUBDEL,bg="black",fg="#00FFFF",font=("Calibri", 16), borderwidth=1,textvariable=CCDEL,insertbackground="#00FFFF")
DELCC_entry.place(x=200,y=170)

def LI(e):
    t4imageL=PhotoImage(file="images/buttons/clear2.png")
    t4buttonL.image=t4imageL
    t4buttonL.config(image=t4imageL,borderwidth=0)
def li(e):
    t4imageL=PhotoImage(file="images/buttons/clear1.png")
    t4buttonL.image=t4imageL
    t4buttonL.config(image=t4imageL,borderwidth=0)
t4imageL=PhotoImage(file="images/buttons/clear1.png")
t4buttonL = tk.Button(tab4, image=t4imageL,bg="black" ,command=tab4clr,activebackground='black',borderwidth=0)
t4buttonL.bind("<Enter>", LI)
t4buttonL.bind("<Leave>", li)

def RI(e):
    t4imageR=PhotoImage(file="images/buttons/search2.png")
    t4buttonR.config(image=t4imageR)
    t4buttonR.image=t4imageR
def ri(e): 
    t4imageR=PhotoImage(file="images/buttons/search1.png")
    t4buttonR.config(image=t4imageR)
    t4buttonR.image=t4imageR
t4imageR=PhotoImage(file="images/buttons/search1.png")
t4buttonR = tk.Button(tab4, image=t4imageR,bg="black",command=lambda:[quit_loop_tab4(),Del_course()],activebackground='black', borderwidth=0)
t4buttonR.bind("<Enter>", RI)
t4buttonR.bind("<Leave>", ri)

t4buttonL.place(x=150,y=250)
t4buttonR.place(x=350,y=250)

t4imagei=PhotoImage(file="images/icons/search.png")
t4buttoni = tk.Button(tab4, image=t2imagei,bg="black",activebackground='black', borderwidth=0)
#t4buttoni.place(x=300,y=120)

############################

###################################### tab 5:functions for attendance ###################################################################################################################################################################################################################################################################################################################################

def clr():
    CCDT_entry.delete(0, END)

def Take_Attendance():

    Year_info=tab5selection

    if Year_info == '1':
        Yr = 'FY'

    elif Year_info == '2':
            Yr = 'SY'

    elif Year_info == '3':
                Yr = 'TY'
    
    # checking the dataset folder is there or not 
    def assure_path_exists(path):
        dir = os.path.dirname(path)
        if not os.path.exists(dir):
            os.makedirs(dir)

    # Create Local Binary Patterns Histograms for face recognization
    recognizer = cv2.face.LBPHFaceRecognizer_create()

    assure_path_exists("trainer/")

    # Load the trained mode
    recognizer.read('trainer/'+Yr+'.yml')

    # Load prebuilt model for Frontal Face
    cascadePath = "haarcascade_frontalface_default.xml"
    date = time.strftime("%d.%m.%Y")
    if not os.path.exists('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info+'/'+'./Attendance/' + date):
        os.makedirs('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info+'/'+'./Attendance/' + date)
    list_of_files = os.listdir('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info+'/'+'./Attendance/' + date)
    LECN_info=(str(len(list_of_files)+1))

    # Create classifier from prebuilt model
    faceCascade = cv2.CascadeClassifier(cascadePath);

    # Set the font style
    font = cv2.FONT_HERSHEY_SIMPLEX

    # Initialize and start the video frame capture
    cam = cv2.VideoCapture(0)

    # Taking the attendance  
    def MarkAttendance(name,Enr):
        date = time.strftime("%d.%m.%Y")
        path ='Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info+'/'+'./Attendance/' + date
        if not os.path.exists(path):
            os.makedirs(path)
        
        def head():
            with open(path+'/'+CC_info+'_'+date+'_Attendance_LEC_'+LECN_info+'.csv','w+') as fp:
                fields = ['Enr.no','Name', 'Time']
                csvwriter = csv.writer(fp)
                csvwriter.writerow(fields)
                
        file_exists = os.path.isfile(path+'/'+CC_info+'_'+date+'_Attendance_LEC_'+LECN_info+'.csv')
        if not file_exists:
                #print("hello")
                head()

        #filename = path + '/Attendance.csv'
        with open(path+'/'+CC_info+'_'+date+'_Attendance_LEC_'+LECN_info+'.csv','r+') as f:
            myData = f.readlines()
            enrs = []
            for line in myData:
                entry = line.split(',')
                enrs.append(entry[0])
            if Enr not in enrs :
                #now =datetime.now()
                dtstr = time.strftime('%H:%M:%S')
                f.writelines(f'\n{Enr},{name},{dtstr}')
                
    # Loop
    while True:
        # Read the video frame
        ret, im =cam.read()

        # Convert the captured frame into grayscale
        gray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)

        # Get all face from the video frame
        faces = faceCascade.detectMultiScale(gray, 1.2,5)

        # For each face in faces
        for(x,y,w,h) in faces:
            path='dataset/'+Yr
            global name
            # Create rectangle around the face
            cv2.rectangle(im, (x-20,y-20), (x+w+20,y+h+20), (0,255,0), 4)

            # Recognize the face belongs to which ID
            Id, confidence = recognizer.predict(gray[y:y+h,x:x+w])
            imagePaths = [os.path.join(path,f) for f in os.listdir(path)]
            for imagePath in imagePaths:
                PIL_img = Image.open(imagePath).convert('L')
                img_numpy = np.array(PIL_img,'uint8')
                ##print(confidence)
                if (confidence<100):
                    if(Id ==int(os.path.split(imagePath)[-1].split(".")[0])):
                        name = str(os.path.split(imagePath)[-1].split(".")[1])
                        Enr = str(os.path.split(imagePath)[-1].split(".")[2])
                else:
                    Id="Unknown"
                    name="Unknown"
                    Enr="Unknown"
                
                ##print(name)
            if(Id!= None) :
                Id = name+" {0:.2f}%".format(round(100 - confidence, 2))
                MarkAttendance(name,Enr)
            else:
                Id = "unknown {0:.2f}%".format(round(100 - confidence, 2))
           
            # Put text describe who is in the picture
            cv2.rectangle(im, (x-22,y-90), (x+w+22, y-22), (0,255,0), -1)
            cv2.putText(im, str(Id), (x,y-40), font, 1, (255,255,255), 3)

        # Display the video frame with the bounded rectangle
        cv2.namedWindow ('Face Rcogniion', cv2.WINDOW_NORMAL)
        cv2.setWindowProperty ('Face Rcogniion', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)
        cv2.imshow ('Face Rcogniion', im)

        # If 'q' is pressed, close program
        if cv2.waitKey(10) & 0xFF == ord('x'):
            break
        
    # Stop the camera
    cam.release()
    Mark_daily()

    # Close all windows
    cv2.destroyAllWindows()

def Mark_daily():
    Year_info=tab5selection

    if Year_info == '1':
        Yr = 'FY'

    elif Year_info == '2':
            Yr = 'SY'

    elif Year_info == '3':
                Yr = 'TY'
                
    list_of_files = os.listdir('STUDENT_DETAILS/'+Yr+'/')
    new_cols = ['ENR','Name','Email','Phone','Present','Absent','Total','Attendance percentage']
    df=pd.DataFrame(index=list_of_files)
    det=[]
    global SNAME,SENRL,SEMAIL,SPHNO
    SNAME=[]
    SENRL=[]
    SEMAIL=[]
    SPHNO=[]
    newdet=[]
    for f in list_of_files:
        with open('STUDENT_DETAILS/'+Yr+'/'+f, 'r') as filehandle:
            for line in filehandle:
                det.append(line)
                ##print(line)
                ##print(det)

        #print(det)
        newdet = [x[:-1] for x in det]
        ##print(newdet)
        ##print(newdet)
        SNAME.append(newdet[0])
        SENRL.append(newdet[1])
        SEMAIL.append(newdet[2])
        SPHNO.append(newdet[3])
        newdet.clear()
        det.clear()
        filehandle.close()
        #newdet.append('done')

    df=df.reindex(df.columns.union(new_cols), axis=1)
    df.ENR = SENRL
    df.Name=SNAME
    df.Email=SEMAIL
    df.Phone=SPHNO
        
    #new_row = ['Unknown','Unknown','Unknown','Unknown',0,0,0,0]
    #df = df.append(pd.Series(new_row, index=df.columns, name='Unknown'))
    df['Present'] = 0
    df['Absent']=0
    df['Total']=0
    df['Attendance percentage']=0
    #print(df)
    ##print(NAME)
    ##print(ENRL)
    ##print(EMAIL)
    ##print(PHNO)

    #####################################################################################################################################3

    global cnt
    cnt=0
    list_of_files = os.listdir('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info+'/'+'./Attendance/')
    #print(list_of_files)
    for new in list_of_files:
        if not os.path.exists('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info+'/'+'./Attendance/'+ new):
            date += timedelta(days=1)
            new=date.strftime("%d.%m.%Y")
            continue
        #print(new)
        list_of_sheets = os.listdir('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info+'/'+'./Attendance/'+ new+'/')
        #print(list_of_sheets)
        for files in list_of_sheets:
            cf = pd.read_csv('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info+'/'+'./Attendance/'+ new+'/'+files)
            matrix2 = cf[cf.columns[0]].to_numpy()
            #print(cf)
            #print(df)
            ##print(type(matrix2))
            list2 = matrix2.tolist()
            #print("the list")
            #print(list2)
            
            for i in list2:
                if i is not None and i!='Unknown':
                    df.at[str(i), 'Present'] +=1
                    #print(i)
                #if i=='Unknown':
                #    df.at[str(i), 'Present'] +=1
                    
            cnt+=1
                
    #print(cnt)
    list_of_files = os.listdir('STUDENT_DETAILS/'+Yr+'/')
    for l in list_of_files:
        #print(l)
        df.at[l, 'Total'] = cnt

    for lf in list_of_files:
        #print(df.at[l, 'Total'] - df.at[lf, 'Present'])
        df.at[lf,'Absent']= (df.at[l, 'Total'] - df.at[lf, 'Present'])

    for af in list_of_files:
        #print(df.at[af, 'Present']/(df.at[af, 'Total']/100))
        df.at[af,'Attendance percentage']= ((df.at[af, 'Present']/(df.at[af, 'Total']/100)))

    #df.at['Unknown', 'Total'] = cnt
    #df.at['Unknown','Absent']= (df.at['Unknown', 'Total'] - df.at['Unknown', 'Present'])
    #df.at['Unknown','Attendance percentage']= ((df.at['Unknown', 'Present']/(df.at['Unknown', 'Total']/100)))

    deflist=[]
    for det in list_of_files:
        #print(d)
        if df.at[det, 'Attendance percentage']< 75:
            deflist.append(det)
                    
    #print(deflist)

    new_cols = ['ENR','Name','Email','Phone']
    cdf=pd.DataFrame(index=deflist)
    cdf=cdf.reindex(cdf.columns.union(new_cols), axis=1)
    #cdf.at['FCOW18136', 'Name'] = 10
    detl=[]
    global DDNAME,DDENRL,DDEMAIL,DDPHNO
    DDNAME=[]
    DDENRL=[]
    DDEMAIL=[]
    DDPHNO=[]
    newdetl=[]
    for f in deflist:
        #print('\n hello  \n')
        with open('STUDENT_DETAILS/'+Yr+'/'+f, 'r') as filehandle:
            for line in filehandle:
                detl.append(line)
                ##print(line)
                ##print(detl)

        #print(detl)
        newdetl = [x[:-1] for x in detl]
        ##print(newdetl)
        ##print(newdetl)
        DDNAME.append(newdetl[0])
        DDENRL.append(newdetl[1])
        DDEMAIL.append(newdetl[2])
        DDPHNO.append(newdetl[3])
        newdetl.clear()
        detl.clear()
        filehandle.close()
        #newdetl.append('done')
        #print('\n hello  \n')

    cdf=cdf.reindex(cdf.columns.union(new_cols), axis=1)
    cdf.ENR = DDENRL
    cdf.Name=DDNAME
    cdf.Email=DDEMAIL
    cdf.Phone=DDPHNO
    
    if not os.path.exists('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info+'/'+'./Regular_Attendance'):
        os.makedirs('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info+'/'+'./Regular_Attendance')
    
    cdf.to_csv('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info+'/'+'./Regular_Attendance/'+CC_info+'_DD.csv', index=False)

    #print(df)
    #print(cdf)
    df.to_csv('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info+'/'+'./Regular_Attendance/'+CC_info+'_Regular_Attendance.csv', index=False)

def Find_course():
    global LAB5
    global CC_info
    CC_info = CCDT.get()
    global LECN_info
    LECN_info = LECN.get()

    Year_info=tab5selection

    if Year_info == '1':
        Yr = 'FY'

    elif Year_info == '2':
            Yr = 'SY'

    elif Year_info == '3':
                Yr = 'TY'
                
    if os.path.exists('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info):
        LAB5=Label(tab5, text="",bg="black", fg="#00FFFF", font=("calibri", 20))
        LAB5.place(x=130,y=500)
        LAB5.configure(text="Details are Founded",width=30)
        CFIND=Frame(tab5,bg="black",highlightbackground="orange",highlightcolor='#24f2e8', highlightthickness=2)
        CFIND.place(x=800,y=80,height=500,width=500)
        tk.Label(CFIND, text="THE FOUNDED DETAILS ",fg="#00ffcc",bg="black",font=("Calibri", 20)).place(x=120,y=5)
        global sh
        sh=[]
        with open('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info+'/'+CC_info, 'r') as filehandle:
            for line in filehandle:
                sh.append(line)
                ##print(line)
        ##print(sh)
        newsh = [x[:-1] for x in sh]
        ##print(newsh)
        Label(CFIND, text="COURSE NAME :",bg="black",fg="white",font=("Calibri", 18), borderwidth=0).place(x=40,y=100)
        Label(CFIND, text=newsh[0],bg="black",fg="white",font=("Calibri", 18), borderwidth=0).place(x=250,y=100)
        Label(CFIND, text="COURSE CODE :",bg="black",fg="white",font=("Calibri", 18), borderwidth=0).place(x=40,y=180)
        Label(CFIND, text=newsh[1],bg="black",fg="white",font=("Calibri", 18), borderwidth=0).place(x=250,y=180)
        Label(CFIND, text="semester:",bg="black",fg="white",font=("Calibri", 18), borderwidth=0).place(x=40,y=260)
        Label(CFIND, text=newsh[2],bg="black",fg="white",font=("Calibri", 18), borderwidth=0).place(x=250,y=260)
        Label(CFIND, text="CREDITS :",bg="black",fg="white",font=("Calibri", 18), borderwidth=0).place(x=40,y=340)
        Label(CFIND, text=newsh[3],bg="black",fg="white",font=("Calibri", 18), borderwidth=0).place(x=250,y=340)

        def LI(e):
            cc2=PhotoImage(file="images/buttons/start2.png")
            Dbutton.image=cc2
            Dbutton.config(image=cc2,borderwidth=0)
        def li(e):
            cc2=PhotoImage(file="images/buttons/start1.png")
            Dbutton.image=cc2
            Dbutton.config(image=cc2,borderwidth=0)

        def IR(e):
            cc1=PhotoImage(file="images/buttons/clear2.png")
            Cbutton.image=cc1
            Cbutton.config(image=cc1,borderwidth=0)
            
        def ir(e):
            cc1=PhotoImage(file="images/buttons/clear1.png")
            Cbutton.image=cc1
            Cbutton.config(image=cc1)
            
        def cancel():
            CFIND.destroy()
            Dbutton.place(x=1000,y=10000)
            Cbutton.place(x=1000,y=10300)

        global cc1
        cc1=PhotoImage(file="images/buttons/clear1.png")
        Cbutton = tk.Button(tab5, image=cc1,command = cancel, borderwidth=0,bg="black",activebackground='black')
        Cbutton.bind("<Enter>", IR)
        Cbutton.bind("<Leave>", ir)
        Cbutton.place(x=840,y=550)

        global cc2
        cc2=PhotoImage(file="images/buttons/start1.png")
        Dbutton = tk.Button(tab5, image=cc2,bg="black" ,command=Take_Attendance,activebackground='black',borderwidth=0)
        Dbutton.bind("<Enter>", LI)
        Dbutton.bind("<Leave>", li)
        Dbutton.place(x=1070,y=550)
 
    else:
        
        LAB5=Label(tab5, text="",bg="black", fg="#00FFFF", font=("RED", 20))
        LAB5.place(x=130,y=500)
        LAB5=Label(tab5, text="The Details are not found ",width=30,bg="black", fg="RED", font=("calibri", 20))
        LAB5.place(x=130,y=500)

######### tab5 starts ################################################################################################################################################################

ftab5=PhotoImage(file = "images/bg/ho.png")
background_label = Label(tab5, image=ftab5)
background_label.place(x=0, y=0, relwidth=1, relheight=1)

tab5VL=Frame(tab5,bg='white')
tab5VL.place(x=700,y=0,height=900,width=2)

tab5MHR=Frame(tab5,bg='white')
tab5MHR.place(x=0,y=340,height=2,width=700)

ATTEND=Frame(tab5,bg="black",highlightbackground="orange",highlightcolor='#24f2e8', highlightthickness=2)
ATTEND.place(x=100,y=10,height=300,width=500)

tk.Label(ATTEND, text="PLEASE ENTER COURSE CODE TO TAKE ATTENDANCE ",fg="#24f2e8",bg="black",font=("Calibri", 16)).place(x=20,y=20)
global CCDT
CCDT = StringVar()
global LECN
LECN = StringVar()

Label(ATTEND, text="Year :",bg="black",fg="white",font=("Calibri", 16), borderwidth=0).place(x=50,y=90)

tab5year = StringVar()
tab5year.set(1)

def quit_loop_tab5():
    #print ("tab5selection:"+tab5year.get())
    global tab5selection
    tab5selection = tab5year.get()

Radiobutton(ATTEND,text="1st year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=tab5year, value = 1).place(x=150,y=90)
Radiobutton(ATTEND,  text="2nd year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=tab5year, value = 2).place(x=250,y=90)
Radiobutton(ATTEND,  text="3rd year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=tab5year, value = 3).place(x=350,y=90)

try:
    tab5selection
except NameError:
    #print("well, it WASN'T defined after all!")
    tab5selection=1
    
Label(ATTEND, text="COURSE CODE :",bg="black",fg="white",font=("Calibri", 16), borderwidth=0).place(x=50,y=170)

CCDT_entry=Entry(ATTEND,bg="black",fg="#00FFFF",font=("Calibri", 16), borderwidth=1,textvariable=CCDT,insertbackground="#00FFFF")
CCDT_entry.place(x=200,y=170)

def LI(e):
    t5imageL=PhotoImage(file="images/buttons/clear2.png")
    t5buttonL.image=t5imageL
    t5buttonL.config(image=t5imageL,borderwidth=0)
def li(e):
    t5imageL=PhotoImage(file="images/buttons/clear1.png")
    t5buttonL.image=t5imageL
    t5buttonL.config(image=t5imageL,borderwidth=0)
t5imageL=PhotoImage(file="images/buttons/clear1.png")
t5buttonL = tk.Button(tab5, image=t5imageL,bg="black" ,command=clr,activebackground='black',borderwidth=0)
t5buttonL.bind("<Enter>", LI)
t5buttonL.bind("<Leave>", li)

def RI(e):
    t5imageR=PhotoImage(file="images/buttons/search2.png")
    t5buttonR.config(image=t5imageR)
    t5buttonR.image=t5imageR
def ri(e): 
    t5imageR=PhotoImage(file="images/buttons/search1.png")
    t5buttonR.config(image=t5imageR)
    t5buttonR.image=t5imageR
t5imageR=PhotoImage(file="images/buttons/search1.png")
t5buttonR = tk.Button(tab5, image=t5imageR,bg="black",command=lambda:[quit_loop_tab5(),Find_course()],activebackground='black', borderwidth=0)
t5buttonR.bind("<Enter>", RI)
t5buttonR.bind("<Leave>", ri)

t5buttonL.place(x=150,y=250)
t5buttonR.place(x=350,y=250)

t5imagei=PhotoImage(file="images/icons/search.png")
t5buttoni = tk.Button(tab5, image=t2imagei,bg="black",activebackground='black', borderwidth=0)
#t5buttoni.place(x=300,y=80)

##############################################################################################################################################################

###################### functions of tab 6 #######################################################################################

def send_attachment():

    if os.path.exists('Teachers/'+userName+'/'+userName):
        global usrdet
        usrdet=[]
        with open('Teachers/'+userName+'/'+userName, 'r') as usrfile:
            for line in usrfile:
                usrdet.append(line)
        newusrdet = [x[:-1] for x in usrdet]
        #print(newusrdet)

    print(userName)
    CC_info = CCDF.get()
    Year_info=tab6selection

    if Year_info == '1':
        Yr = 'FY'

    elif Year_info == '2':
            Yr = 'SY'

    elif Year_info == '3':
                Yr = 'TY'
    serverlist=[CC_info]
    print(newusrdet[1])
    print(newusrdet[1])
    print(serverlist)
    emailfrom = "attendancesystem36@gmail.com"
    emailto = newusrdet[1]
    fileToSend = 'Teachers/'+userName+'/COURSES/'+Yr+'/'+CCFD+'/'+'./Defaulters/Defaulters.csv'
    username = "attendancesystem36@gmail.com"
    password = "Attendance_2020"

    msg = MIMEMultipart()
    msg["From"] = emailfrom
    msg["To"] = emailto
    msg["Subject"] = "The Defaulters list "
    msg.preamble = "The defulters list "

    ctype, encoding = mimetypes.guess_type(fileToSend)
    if ctype is None or encoding is not None:
        ctype = "application/octet-stream"

    maintype, subtype = ctype.split("/", 1)

    if maintype == "text":
        fp = open(fileToSend)
        # Note: we should handle calculating the charset
        attachment = MIMEText(fp.read(), _subtype=subtype)
        fp.close()
    elif maintype == "image":
        fp = open(fileToSend, "rb")
        attachment = MIMEImage(fp.read(), _subtype=subtype)
        fp.close()
    elif maintype == "audio":
        fp = open(fileToSend, "rb")
        attachment = MIMEAudio(fp.read(), _subtype=subtype)
        fp.close()
    else:
        fp = open(fileToSend, "rb")
        attachment = MIMEBase(maintype, subtype)
        attachment.set_payload(fp.read())
        fp.close()
        encoders.encode_base64(attachment)
    attachment.add_header("Content-Disposition", "attachment", filename=fileToSend)
    msg.attach(attachment)

    serverlist[0] = smtplib.SMTP("smtp.gmail.com:587")
    serverlist[0].starttls()
    serverlist[0].login(username,password)
    serverlist[0].sendmail(emailfrom, emailto, msg.as_string())
    serverlist[0].quit()
    emailfrom = "attendancesystem36@gmail.com"
    emailto = "jainam.purushotam@btech.christuniversity.in"
    fileToSend = 'Teachers/'+userName+'/COURSES/'+Yr+'/'+CCFD+'/'+'./Defaulters/Defaulters.csv'
    username = "attendancesystem36@gmail.com"
    password = "Attendance_2020"

    msg = MIMEMultipart()
    msg["From"] = emailfrom
    msg["To"] = emailto
    msg["Subject"] = "The Defaulters list "
    msg["Body"]= "Defaulters list for this course"
    msg.preamble = "The defulters list "

    ctype, encoding = mimetypes.guess_type(fileToSend)
    if ctype is None or encoding is not None:
        ctype = "application/octet-stream"

    maintype, subtype = ctype.split("/", 1)

    if maintype == "text":
        fp = open(fileToSend)
        # Note: we should handle calculating the charset
        attachment = MIMEText(fp.read(), _subtype=subtype)
        fp.close()
    elif maintype == "image":
        fp = open(fileToSend, "rb")
        attachment = MIMEImage(fp.read(), _subtype=subtype)
        fp.close()
    elif maintype == "audio":
        fp = open(fileToSend, "rb")
        attachment = MIMEAudio(fp.read(), _subtype=subtype)
        fp.close()
    else:
        fp = open(fileToSend, "rb")
        attachment = MIMEBase(maintype, subtype)
        attachment.set_payload(fp.read())
        fp.close()
        encoders.encode_base64(attachment)
    attachment.add_header("Content-Disposition", "attachment", filename=fileToSend)
    msg.attach(attachment)

    serverlist[0] = smtplib.SMTP("smtp.gmail.com:587")
    serverlist[0].starttls()
    serverlist[0].login(username,password)
    serverlist[0].sendmail(emailfrom, emailto, msg.as_string())
    serverlist[0].quit()

def Cal_def():
    CC_info = CCDF.get()
    Year_info=tab6selection

    if Year_info == '1':
        Yr = 'FY'

    elif Year_info == '2':
            Yr = 'SY'

    elif Year_info == '3':
                Yr = 'TY'

    if not os.path.exists('A1123.csv'):
        list_of_files = os.listdir('STUDENT_DETAILS/'+Yr)
        new_cols = ['ENR','Name','Email','Phone','Present','Absent','Total','Attendance percentage']
        df=pd.DataFrame(index=list_of_files)
        det=[]
        global NAME,ENRL,EMAIL,PHNO
        NAME=[]
        ENRL=[]
        EMAIL=[]
        PHNO=[]
        newdet=[]
        for f in list_of_files:
            #print('\n hello  \n')
            with open('STUDENT_DETAILS/'+Yr+'/'+f, 'r') as filehandle:
                for line in filehandle:
                    det.append(line)
                    ##print(line)
                    ##print(det)

            #print(det)
            newdet = [x[:-1] for x in det]
            ##print(newdet)
            ##print(newdet)
            NAME.append(newdet[0])
            ENRL.append(newdet[1])
            EMAIL.append(newdet[2])
            PHNO.append(newdet[3])
            newdet.clear()
            det.clear()
            filehandle.close()
            #newdet.append('done')
            #print('\n hello  \n')

        df=df.reindex(df.columns.union(new_cols), axis=1)
        df.ENR = ENRL
        df.Name=NAME
        df.Email=EMAIL
        df.Phone=PHNO
        
    #new_row = ['Unknown','Unknown','Unknown','Unknown',0,0,0,0]
    #df = df.append(pd.Series(new_row, index=df.columns, name='Unknown'))
    df['Present'] = 0
    df['Absent']=0
    df['Total']=0
    df['Attendance percentage']=0
    #print("checkig df")
    #print(df)
    ##print(NAME)
    ##print(ENRL)
    ##print(EMAIL)
    ##print(PHNO)

    #####################################################################################################################################3


    date = datetime.datetime(int(y.get()),int(m.get()),int(d.get()))

    d0 = datetime.datetime(int(y.get()),int(m.get()),int(d.get()))
    d1 = datetime.datetime(int(yt.get()),int(mt.get()),int(dt.get()))
    delta = d1 - d0
    ##print(delta.days)
    global count
    count=0
    new=date.strftime("%d.%m.%Y")
    #count=1
    for i in range((delta.days)+1):
        if not os.path.exists('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info+'/'+'./Attendance/'+ new):
            #print('terminated')
            date += timedelta(days=1)
            new=date.strftime("%d.%m.%Y")
            continue
        #print(new)
        list_of_files = os.listdir('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info+'/'+'./Attendance/'+ new+'/')
        for files in list_of_files:
            cf = pd.read_csv('Teachers/'+userName+'/COURSES/'+Yr+'/'+CC_info+'/'+'./Attendance/'+ new+'/'+files)
            matrix2 = cf[cf.columns[0]].to_numpy()
            ##print(type(matrix2))
            list2 = matrix2.tolist()
            #print(list2)
            for i in list2:
                if i is not None and i!='Unknown':
                    df.at[i, 'Present'] +=1
                    #print(i)
                #if i=='Unknown':
                 #   df.at[i, 'Present'] +=1
                    
            count+=1
            
        date += timedelta(days=1)
        new=date.strftime("%d.%m.%Y")
            #count++
    #print(count)
    list_of_files = os.listdir('STUDENT_DETAILS/'+Yr)
    for l in list_of_files:
        #print(l)
        df.at[l, 'Total'] = count

    for lf in list_of_files:
        #print(df.at[l, 'Total'] - df.at[lf, 'Present'])
        df.at[lf,'Absent']= (df.at[l, 'Total'] - df.at[lf, 'Present'])

    for af in list_of_files:
        #print(df.at[af, 'Present']/(df.at[af, 'Total']/100))
        df.at[af,'Attendance percentage']= ((df.at[af, 'Present']/(df.at[af, 'Total']/100)))

    #df.at['Unknown', 'Total'] = count
    #df.at['Unknown','Absent']= (df.at['Unknown', 'Total'] - df.at['Unknown', 'Present'])
    #df.at['Unknown','Attendance percentage']= ((df.at['Unknown', 'Present']/(df.at['Unknown', 'Total']/100)))
        
    Def_df = df[df['Attendance percentage'] < 75]
    print('Defaulters dataframe')
    print(Def_df)

    deflist=[]
    for det in list_of_files:
        #print(d)
        if df.at[det, 'Attendance percentage']< 75:
            deflist.append(det)
            
    if not os.path.exists('Teachers/'+userName+'/COURSES/'+Yr+'/'+CCFD+'/'+'./Defaulters'):
        os.makedirs('Teachers/'+userName+'/COURSES/'+Yr+'/'+CCFD+'/'+'./Defaulters')
    

    Def_df.to_csv('Teachers/'+userName+'/COURSES/'+Yr+'/'+CCFD+'/'+'./Defaulters/Defaulters.csv', index=False)

    #print(df)
    #print(cdf)
    df.to_csv('Teachers/'+userName+'/COURSES/'+Yr+'/'+CCFD+'/'+'./Defaulters/Attendance.csv', index=False)
    
    count_row = Def_df.shape[0]
    
    DEFA=Frame(tab6,bg="black",highlightbackground="yellow", highlightthickness=2)
    DEFA.place(x=820,y=50,height=500,width=500)
    Label(DEFA, text="THE STUDENT DETAILS ",bg="black",fg="aqua",font=("Calibri", 16), borderwidth=0).place(x=130,y=10)
    Label(DEFA, text="STUDENT NAME",bg="black",fg="red",font=("Calibri", 14,'bold'), borderwidth=0).place(x=10,y=60)
    Label(DEFA, text="ENR_NO",bg="black",fg="red",font=("Calibri", 14,'bold'), borderwidth=0).place(x=200,y=60)
    Label(DEFA, text="ATTEND.",bg="black",fg="red",font=("Calibri", 14,'bold'), borderwidth=0).place(x=350,y=60)
    yx=100
    for en in deflist:
        Label(DEFA, text=Def_df.at[en,'Name'],bg="black",fg="white",font=("Calibri", 14,'bold'), borderwidth=0).place(x=10,y=yx)
        Label(DEFA, text=Def_df.at[en,'ENR'],bg="black",fg="white",font=("Calibri", 14,'bold'), borderwidth=0).place(x=200,y=yx)
        Label(DEFA, text=Def_df.at[en,'Attendance percentage'],bg="black",fg="white",font=("Calibri", 14,'bold'), borderwidth=0).place(x=370,y=yx)
        yx=yx+30

    s=[]
    def send_mail():
        i=0
        for e in deflist :
            
            # creates SMTP session 
            deflist[i] = smtplib.SMTP('smtp.gmail.com', 587) 
              
            # start TLS for security 
            deflist[i].starttls() 
              
            # Authentication 
            deflist[i].login("attendancesystem36@gmail.com", "Attendance_2020") 
              
            # message to be sent
            message = """Subject: Defaulters Alert \n
            Hey,this is an important message \n your attendance is very low so you are in defaulters list """
            
            #message = "Your child is in defaulters list"
              
            # sending the mail 
            deflist[i].sendmail("attendancesystem36@gmail.com", Def_df.at[en,'Email'], message) 
            
            print(Def_df.at[en,'Email'])
            
            # terminating the session 
            deflist[i].quit()

            i=i+1
        print("Successfull")

    def LI(e):
        cl2=PhotoImage(file="images/buttons/send-mail2.png")
        Dbutton.image=cl2
        Dbutton.config(image=cl2,borderwidth=0)
    def li(e):
        cl2=PhotoImage(file="images/buttons/send-mail1.png")
        Dbutton.image=cl2
        Dbutton.config(image=cl2,borderwidth=0)

    def IR(e):
        cl1=PhotoImage(file="images/buttons/clear2.png")
        Cbutton.image=cl1
        Cbutton.config(image=cl1,borderwidth=0)
        
    def ir(e):
        cl1=PhotoImage(file="images/buttons/clear1.png")
        Cbutton.image=cl1
        Cbutton.config(image=cl1)
        
    def cancel():
        DEFA.destroy()
        Dbutton.place(x=1000,y=10000)
        Cbutton.place(x=1000,y=10300)
    
    global cl1
    cl1=PhotoImage(file="images/buttons/clear1.png")
    Cbutton = tk.Button(DEFA, image=cl1,command = cancel, borderwidth=0,bg="black",activebackground='black')
    Cbutton.bind("<Enter>", IR)
    Cbutton.bind("<Leave>", ir)
    Cbutton.place(x=30,y=400)

    global cl2
    cl2=PhotoImage(file="images/buttons/send-mail1.png")
    Dbutton = tk.Button(DEFA, image=cl2,bg="black" ,command=send_mail,activebackground='black',borderwidth=0)
    Dbutton.bind("<Enter>", LI)
    Dbutton.bind("<Leave>", li)
    Dbutton.place(x=280,y=400)

    send_attachment()
        
def Search_date():    
    global fdt
    global tdt

    Year_info=tab6selection

    if Year_info == '1':
        Yr = 'FY'

    elif Year_info == '2':
            Yr = 'SY'

    elif Year_info == '3':
                Yr = 'TY'

    #fdt=str(datetime.datetime(int(y.get()),int(m.get()),int(d.get())))
    fdt=(d.get()+'.'+m.get()+'.'+y.get())
    ##print(fdt)

    tdt=(dt.get()+'.'+mt.get()+'.'+yt.get())
    
    ##print(tdt)
    #if os.path.exists('Teachers/'+userName+'/COURSES/'+CC_info+'/'+'./Attendance/' + fdt):
    #function for trying to get correct date not used 
    correctDate = None
    try:
        newDate = datetime.datetime(int(y.get()),int(m.get()),int(d.get()))
        correctDate = True
    except ValueError:
        correctDate = False
    ##print(str(correctDate))

    if not os.path.exists('Teachers/'+userName+'/COURSES/'+Yr+'/'+CCFD+'/'+'./Attendance/' + fdt):
        LAB6=Label(tab6, text="",bg="black", fg="red", font=("calibri", 20))
        LAB6.place(x=150,y=600)
        LAB6.configure(text="The starting date is not found ",width=30)
        
    elif not os.path.exists('Teachers/'+userName+'/COURSES/'+Yr+'/'+CCFD+'/'+'./Attendance/' + tdt):
        LAB6=Label(tab6, text="",bg="black", fg="#00FFFF", font=("RED", 20))
        LAB6.place(x=150,y=600)
        LAB6=Label(tab6, text="The Ending date is not found ",width=30,bg="black", fg="RED", font=("calibri", 20))
        LAB6.place(x=150,y=600)

    else:
        LAB6=Label(tab6, text="",bg="black", fg="#00FFFF", font=("RED", 20))
        LAB6.place(x=150,y=600)
        LAB6=Label(tab6, text="The deatils are found ",width=30,bg="black", fg="#00FFFF", font=("calibri", 20))
        LAB6.place(x=150,y=600)
        Cal_def()

def Search_course():
    global LAB6
    global CCFD
    
    Year_info=tab6selection

    if Year_info == '1':
        Yr = 'FY'

    elif Year_info == '2':
        Yr = 'SY'

    elif Year_info == '3':
        Yr = 'TY'

    CCFD = CCDF.get()

    if os.path.exists('Teachers/'+userName+'/COURSES/'+Yr+'/'+CCFD):
        LAB6=Label(tab6, text="",bg="black", fg="#00FFFF", font=("calibri", 20))
        LAB6.place(x=150,y=600)
        LAB6.configure(text="Details are Founded",width=30)
        Search_date()

    else:
        
        LAB6=Label(tab6, text="",bg="black", fg="#00FFFF", font=("RED", 20))
        LAB6.place(x=150,y=600)
        LAB6=Label(tab6, text="The Course is not found ",width=30,bg="black", fg="RED", font=("calibri", 20))
        LAB6.place(x=150,y=600)

################################################# tab6 start #############################################################################################################

ftab6=PhotoImage(file = "images/bg/ho.png")
background_label = Label(tab6, image=ftab6)
background_label.place(x=0, y=0, relwidth=1, relheight=1)

var = IntVar()
CADE=Frame(tab6,bg="#000000",highlightbackground="orange",highlightcolor='#24f2e8', highlightthickness=2)
CADE.place(x=10,y=150,height=350,width=750)

VRM=Frame(tab6,bg='white')
VRM.place(x=800,y=0,height=900,width=2)

tk.Label(CADE, text="PLEASE ENTER DETAILS TO CALCULATE DEAFAULTER ",fg="#24f2e8",bg="#000000",font=("Calibri", 20)).place(x=70,y=60)

Label(CADE, text="Course code :",bg="#000000",fg="white",font=("Calibri", 18), borderwidth=0).place(x=90,y=130)

global CCDF
CCDF = StringVar()
global d,m,y
d = StringVar()
m = StringVar()
y = StringVar()

global dt,mt,yt
dt = StringVar()
mt = StringVar()
yt = StringVar()

DCOD_entry=Entry(CADE,bg="#000000",fg="#FFFFFF",textvariable=CCDF,insertbackground="#00FFFF",font=("Calibri", 15), borderwidth=1)
DCOD_entry.place(x=260,y=130)

Label(CADE, text="Duration :",bg="#000000",fg="white",font=("Calibri", 18), borderwidth=0).place(x=90,y=200)

DE = Spinbox(CADE, from_ = 1, to = 31,bg="#000000",textvariable=d,insertbackground="#00FFFF",fg="#FFFFFF",font=("Calibri", 15))
ME= Spinbox(CADE, from_ = 1, to = 12,bg="#000000",textvariable=m,insertbackground="#00FFFF",fg="#FFFFFF",font=("Calibri", 15))
YE = Spinbox(CADE, from_ = 2000, to = 2100,bg="#000000",textvariable=y,insertbackground="#00FFFF",fg="#FFFFFF",font=("Calibri", 15))
DE.place(x=260,y=200,width=50)
ME.place(x=310,y=200,width=50)
YE.place(x=360,y=200,width=100)

Label(CADE, text=" To ",bg="#000000",fg="white",font=("Calibri", 18), borderwidth=0).place(x=465,y=200)

DE1 = Spinbox(CADE, from_ = 1, to = 31,bg="#000000",insertbackground="#00FFFF",fg="#FFFFFF",textvariable=dt,font=("Calibri", 15))
ME1 = Spinbox(CADE, from_ = 1, to = 12,bg="#000000",insertbackground="#00FFFF",fg="#FFFFFF",textvariable=mt,font=("Calibri", 15))
YE1 = Spinbox(CADE, from_ = 2000, to = 2100,bg="#000000",insertbackground="#00FFFF",fg="#FFFFFF",textvariable=yt,font=("Calibri", 15))
DE1.place(x=510,y=200,width=50)
ME1.place(x=560,y=200,width=50)
YE1.place(x=610,y=200,width=100)

Label(CADE, text="Year :",bg="black",fg="white",font=("Calibri", 16), borderwidth=0).place(x=90,y=260)

tab6year = StringVar()
tab6year.set(1)

def quit_loop_tab6():
    #print ("tab6selection:"+tab6year.get())
    global tab6selection
    tab6selection = tab6year.get()

Radiobutton(CADE,text="1st year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=tab6year, value = 1).place(x=260,y=260)
Radiobutton(CADE,  text="2nd year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=tab6year, value = 2).place(x=360,y=260)
Radiobutton(CADE,  text="3rd year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=tab6year, value = 3).place(x=460,y=260)

try:
    tab6selection
except NameError:
    #print("well, it WASN'T defined after all!")
    tab6selection=1

def dI(e):
    dimage=PhotoImage(file="images/buttons/cal.png")
    dbutton.config(image=dimage)
    dbutton.image=dimage
def Di(e):
    dimage=PhotoImage(file="images/buttons/cal1.png")
    dbutton.config(image=dimage)
    dbutton.image=dimage
dimage=PhotoImage(file="images/buttons/cal1.png")
dbutton = tk.Button(tab6, image=dimage,command=lambda:[quit_loop_tab6(),Search_course()],bg="#000000",activebackground='#000000', borderwidth=0)
dbutton.bind("<Enter>", dI)
dbutton.bind("<Leave>", Di)

dbutton.place(x=300,y=475)

t6imagei=PhotoImage(file="images/icons/attend.png")
t6buttoni = tk.Button(tab6, image=t6imagei,bg="black",activebackground='black', borderwidth=0)
t6buttoni.place(x=350,y=100)

###########################################################################################################################################################

#############################################  Tab7 functions ##############################################################################################################

def send_det_attachment():

    if os.path.exists('Teachers/'+userName+'/'+userName):
        global usrdet
        usrdet=[]
        with open('Teachers/'+userName+'/'+userName, 'r') as usrfile:
            for line in usrfile:
                usrdet.append(line)
        newusrdet = [x[:-1] for x in usrdet]
        #print(newusrdet)

    print(userName)
    CCDTNINFO = CCDTN.get()
    Year_info=tab7selection
    
    if Year_info == '1':
        Yr = 'FY'

    elif Year_info == '2':
            Yr = 'SY'

    elif Year_info == '3':
                Yr = 'TY'
                
    serverlist=[CCDTNINFO]
    print(newusrdet[1])
    print(newusrdet[1])
    print(serverlist)
    emailfrom = "attendancesystem36@gmail.com"
    emailto = newusrdet[1]
    fileToSend = 'Teachers/'+userName+'/COURSES/'+Yr+'/'+CCDTNINFO+'/'+'./Final_Attendance/Detained.csv'
    username = "attendancesystem36@gmail.com"
    password = "Attendance_2020"

    msg = MIMEMultipart()
    msg["From"] = emailfrom
    msg["To"] = emailto
    msg["Subject"] = "The Detaintion list "
    msg.preamble = "The Detaintion list "

    ctype, encoding = mimetypes.guess_type(fileToSend)
    if ctype is None or encoding is not None:
        ctype = "application/octet-stream"

    maintype, subtype = ctype.split("/", 1)

    if maintype == "text":
        fp = open(fileToSend)
        # Note: we should handle calculating the charset
        attachment = MIMEText(fp.read(), _subtype=subtype)
        fp.close()
    elif maintype == "image":
        fp = open(fileToSend, "rb")
        attachment = MIMEImage(fp.read(), _subtype=subtype)
        fp.close()
    elif maintype == "audio":
        fp = open(fileToSend, "rb")
        attachment = MIMEAudio(fp.read(), _subtype=subtype)
        fp.close()
    else:
        fp = open(fileToSend, "rb")
        attachment = MIMEBase(maintype, subtype)
        attachment.set_payload(fp.read())
        fp.close()
        encoders.encode_base64(attachment)
    attachment.add_header("Content-Disposition", "attachment", filename=fileToSend)
    msg.attach(attachment)

    serverlist[0] = smtplib.SMTP("smtp.gmail.com:587")
    serverlist[0].starttls()
    serverlist[0].login(username,password)
    serverlist[0].sendmail(emailfrom, emailto, msg.as_string())
    serverlist[0].quit()
    emailfrom = "attendancesystem36@gmail.com"
    emailto = "jainam.purushotam@btech.christuniversity.in"
    fileToSend = 'Teachers/'+userName+'/COURSES/'+Yr+'/'+CCDTNINFO+'/'+'./Final_Attendance/Detained.csv'
    username = "attendancesystem36@gmail.com"
    password = "Attendance_2020"

    msg = MIMEMultipart()
    msg["From"] = emailfrom
    msg["To"] = emailto
    msg["Subject"] = "The Detaintion list "
    msg["Body"]= "Detained students list for this course"
    msg.preamble = "The Detaintion list of course "+CCDTNINFO+" of "+Yr+"."

    ctype, encoding = mimetypes.guess_type(fileToSend)
    if ctype is None or encoding is not None:
        ctype = "application/octet-stream"

    maintype, subtype = ctype.split("/", 1)

    if maintype == "text":
        fp = open(fileToSend)
        # Note: we should handle calculating the charset
        attachment = MIMEText(fp.read(), _subtype=subtype)
        fp.close()
    elif maintype == "image":
        fp = open(fileToSend, "rb")
        attachment = MIMEImage(fp.read(), _subtype=subtype)
        fp.close()
    elif maintype == "audio":
        fp = open(fileToSend, "rb")
        attachment = MIMEAudio(fp.read(), _subtype=subtype)
        fp.close()
    else:
        fp = open(fileToSend, "rb")
        attachment = MIMEBase(maintype, subtype)
        attachment.set_payload(fp.read())
        fp.close()
        encoders.encode_base64(attachment)
    attachment.add_header("Content-Disposition", "attachment", filename=fileToSend)
    msg.attach(attachment)

    serverlist[0] = smtplib.SMTP("smtp.gmail.com:587")
    serverlist[0].starttls()
    serverlist[0].login(username,password)
    serverlist[0].sendmail(emailfrom, emailto, msg.as_string())
    serverlist[0].quit()

def Cal_det():

    Year_info=tab7selection

    if Year_info == '1':
        Yr = 'FY'

    elif Year_info == '2':
            Yr = 'SY'

    elif Year_info == '3':
                Yr = 'TY'
    
    list_of_files = os.listdir('STUDENT_DETAILS/'+Yr)
    new_cols = ['ENR','Name','Email','Phone','Present','Absent','Total','Attendance percentage']
    df=pd.DataFrame(index=list_of_files)
    det=[]
    global NAME,ENRL,EMAIL,PHNO
    NAME=[]
    ENRL=[]
    EMAIL=[]
    PHNO=[]
    newdet=[]
    for f in list_of_files:
        #print('\n hello  \n')
        with open('STUDENT_DETAILS/'+Yr+'/'+f, 'r') as filehandle:
            for line in filehandle:
                det.append(line)
                ##print(line)
                ##print(det)

        #print(det)
        newdet = [x[:-1] for x in det]
        ##print(newdet)
        ##print(newdet)
        NAME.append(newdet[0])
        ENRL.append(newdet[1])
        EMAIL.append(newdet[2])
        PHNO.append(newdet[3])
        newdet.clear()
        det.clear()
        filehandle.close()
        #newdet.append('done')
        #print('\n hello  \n')

    df=df.reindex(df.columns.union(new_cols), axis=1)
    df.ENR = ENRL
    df.Name=NAME
    df.Email=EMAIL
    df.Phone=PHNO
        
    #new_row = ['Unknown','Unknown','Unknown','Unknown',0,0,0,0]
    #df = df.append(pd.Series(new_row, index=df.columns, name='Unknown'))
    df['Present'] = 0
    df['Absent']=0
    df['Total']=0
    df['Attendance percentage']=0
    #print(df)
    ##print(NAME)
    ##print(ENRL)
    ##print(EMAIL)
    ##print(PHNO)

    #####################################################################################################################################3

    global count
    count=0
    list_of_files = os.listdir('Teachers/'+userName+'/COURSES/'+Yr+'/'+CCDTNINFO+'/'+'./Attendance/')
    for new in list_of_files:
        if not os.path.exists('Teachers/'+userName+'/COURSES/'+Yr+'/'+CCDTNINFO+'/'+'./Attendance/'+ new):
            #print('terminated')
            date += timedelta(days=1)
            new=date.strftime("%d.%m.%Y")
            continue
        #if new not in list_of_files:
        #    #print('break')
        #print(new)
        list_of_sheets = os.listdir('Teachers/'+userName+'/COURSES/'+Yr+'/'+CCDTNINFO+'/'+'./Attendance/'+ new+'/')
        ##print(list_of_files)
        for files in list_of_sheets:
        ##print(iris.head())
            cf = pd.read_csv('Teachers/'+userName+'/COURSES/'+Yr+'/'+CCDTNINFO+'/'+'./Attendance/'+ new+'/'+files)
            matrix2 = cf[cf.columns[0]].to_numpy()
            ##print(type(matrix2))
            list2 = matrix2.tolist()
            #print(list2)
            for i in list2:
                if i is not None and i!='Unknown':
                    df.at[i, 'Present'] +=1
                    #print(i)
                #if i=='Unknown':
                #    df.at[i, 'Present'] +=1
                    
            count+=1
                
            #count++
    #print(count)
    list_of_files = os.listdir('STUDENT_DETAILS/'+Yr)
    for l in list_of_files:
        #print(l)
        df.at[l, 'Total'] = count

    for lf in list_of_files:
        #print(df.at[l, 'Total'] - df.at[lf, 'Present'])
        df.at[lf,'Absent']= (df.at[l, 'Total'] - df.at[lf, 'Present'])

    for af in list_of_files:
        #print(df.at[af, 'Present']/(df.at[af, 'Total']/100))
        df.at[af,'Attendance percentage']= ((df.at[af, 'Present']/(df.at[af, 'Total']/100)))

    #df.at['Unknown', 'Total'] = count
    #df.at['Unknown','Absent']= (df.at['Unknown', 'Total'] - df.at['Unknown', 'Present'])
    #df.at['Unknown','Attendance percentage']= ((df.at['Unknown', 'Present']/(df.at['Unknown', 'Total']/100)))

    Det_df = df[df['Attendance percentage'] < 75]
    print('Defaulters dataframe')
    print(Det_df)

    deflist=[]
    for det in list_of_files:
        #print(d)
        if df.at[det, 'Attendance percentage']< 75:
            deflist.append(det)

    if not os.path.exists('Teachers/'+userName+'/COURSES/'+Yr+'/'+CCDTNINFO+'/'+'./Final_Attendance'):
        os.makedirs('Teachers/'+userName+'/COURSES/'+Yr+'/'+CCDTNINFO+'/'+'./Final_Attendance')
    
    Det_df.to_csv('Teachers/'+userName+'/COURSES/'+Yr+'/'+CCDTNINFO+'/'+'./Final_Attendance/Detained.csv', index=False)

    #print(df)
    #print(cdf)
    df.to_csv('Teachers/'+userName+'/COURSES/'+Yr+'/'+CCDTNINFO+'/'+'./Final_Attendance/Final_Attendance.csv', index=False)

    count_row = Det_df.shape[0]
    
    DETA=Frame(tab7,bg="black",highlightbackground="yellow", highlightthickness=2)
    DETA.place(x=820,y=50,height=500,width=500)
    Label(DETA, text="THE STUDENT DETAILS ",bg="black",fg="aqua",font=("Calibri", 16), borderwidth=0).place(x=130,y=10)
    Label(DETA, text="STUDENT NAME",bg="black",fg="red",font=("Calibri", 14,'bold'), borderwidth=0).place(x=10,y=60)
    Label(DETA, text="ENR_NO",bg="black",fg="red",font=("Calibri", 14,'bold'), borderwidth=0).place(x=200,y=60)
    Label(DETA, text="ATTEND.",bg="black",fg="red",font=("Calibri", 14,'bold'), borderwidth=0).place(x=350,y=60)
    yx=100
    for en in deflist:
        Label(DETA, text=Det_df.at[en,'Name'],bg="black",fg="white",font=("Calibri", 14,'bold'), borderwidth=0).place(x=10,y=yx)
        Label(DETA, text=Det_df.at[en,'ENR'],bg="black",fg="white",font=("Calibri", 14,'bold'), borderwidth=0).place(x=200,y=yx)
        Label(DETA, text=Det_df.at[en,'Attendance percentage'],bg="black",fg="white",font=("Calibri", 14,'bold'), borderwidth=0).place(x=370,y=yx)
        yx=yx+30

    s=[]
    def send_mail():
        CCDTNINFO = CCDTN.get()
        i=0
        for e in deflist :
            
            # creates SMTP session 
            deflist[i] = smtplib.SMTP('smtp.gmail.com', 587) 
              
            # start TLS for security 
            deflist[i].starttls() 
              
            # Authentication 
            deflist[i].login("attendancesystem36@gmail.com", "Attendance_2020") 
              
            # message to be sent
            message = """Subject: Detaintion Alert \n
            Hey,This is an important message you are Detained in  """+CCDTNINFO+""" COURESE """
                
            #message = "Your child is in DETAulters list"
              
            # sending the mail 
            deflist[i].sendmail("attendancesystem36@gmail.com", Det_df.at[en,'Email'], message) 
            
            print(Det_df.at[en,'Email'])
            
            # terminating the session 
            deflist[i].quit()

            i=i+1
        print("Successfull")

    def LI(e):
        cdt2=PhotoImage(file="images/buttons/send-mail2.png")
        Dbutton.image=cdt2
        Dbutton.config(image=cdt2,borderwidth=0)
    def li(e):
        cdt2=PhotoImage(file="images/buttons/send-mail1.png")
        Dbutton.image=cdt2
        Dbutton.config(image=cdt2,borderwidth=0)

    def IR(e):
        cdt1=PhotoImage(file="images/buttons/clear2.png")
        Cbutton.image=cdt1
        Cbutton.config(image=cdt1,borderwidth=0)
        
    def ir(e):
        cdt1=PhotoImage(file="images/buttons/clear1.png")
        Cbutton.image=cdt1
        Cbutton.config(image=cdt1)
          
    def cancel():
        DETA.destroy()
        Dbutton.place(x=1000,y=10000)
        Cbutton.place(x=1000,y=10300)
        
    global cdt1
    cdt1=PhotoImage(file="images/buttons/clear1.png")
    Cbutton = tk.Button(DETA, image=cdt1,command = cancel, borderwidth=0,bg="black",activebackground='black')
    Cbutton.bind("<Enter>", IR)
    Cbutton.bind("<Leave>", ir)
    Cbutton.place(x=30,y=400)

    global cdt2
    cdt2=PhotoImage(file="images/buttons/send-mail1.png")
    Dbutton = tk.Button(DETA, image=cdt2,bg="black" ,command=send_mail,activebackground='black',borderwidth=0)
    Dbutton.bind("<Enter>", LI)
    Dbutton.bind("<Leave>", li)
    Dbutton.place(x=280,y=400)

    send_det_attachment()

def Search_cdet():
    global LAB7
    global CCDTNINFO
    Year_info=tab7selection

    if Year_info == '1':
        Yr = 'FY'

    elif Year_info == '2':
            Yr = 'SY'

    elif Year_info == '3':
                Yr = 'TY'
    CCDTNINFO = CCDTN.get()

    #list_of_files = os.listdir('Teachers/'+userName+'/COURSES/'+CCDTNINFO+'/'+'./Attendance/')
    ##print(list_of_files)
    
    if os.path.exists('Teachers/'+userName+'/COURSES/'+Yr+'/'+CCDTNINFO):
        LAB7=Label(tab7, text="",bg="black", fg="#00FFFF", font=("calibri", 20))
        LAB7.place(x=130,y=600)
        LAB7.configure(text="Details are Founded",width=30)
        Cal_det()
 
    else:
        
        LAB7=Label(tab7, text="",bg="black", fg="#00FFFF", font=("RED", 20))
        LAB7.place(x=130,y=500)
        LAB7=Label(tab7, text="The Course is not found ",width=30,bg="black", fg="RED", font=("calibri", 20))
        LAB7.place(x=130,y=600)

############### tab7 starts  ##################################################################################################################

ftab7=PhotoImage(file = "images/bg/ho.png")
background_label = Label(tab7, image=ftab7)
background_label.place(x=0, y=0, relwidth=1, relheight=1)

var = IntVar()
CADT=Frame(tab7,bg="#000000",highlightbackground="orange",highlightcolor='#24f2e8', highlightthickness=2)
CADT.place(x=10,y=150,height=350,width=750)

VM=Frame(tab7,bg='white')
VM.place(x=800,y=0,height=900,width=2)

tk.Label(CADT, text="PLEASE ENTER DETAILS TO CALCULATE DETENTION ",fg="#24f2e8",bg="#000000",font=("Calibri", 20)).place(x=70,y=60)

Label(CADT, text="Year :",bg="black",fg="white",font=("Calibri", 16), borderwidth=0).place(x=90,y=130)

tab7year = StringVar()
tab7year.set(1)

def quit_loop_tab7():
    #print ("tab7selection:"+tab7year.get())
    global tab7selection
    tab7selection = tab7year.get()

Radiobutton(CADT,text="1st year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=tab7year, value = 1).place(x=260,y=130)
Radiobutton(CADT,  text="2nd year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=tab7year, value = 2).place(x=360,y=130)
Radiobutton(CADT,  text="3rd year",font=("Calibri", 16),bg="black",fg="white",activeforeground='white', selectcolor='black',activebackground='black', variable=tab7year, value = 3).place(x=460,y=130)

try:
    tab7selection
except NameError:
    #print("well, it WASN'T defined after all!")
    tab7selection=1
    
Label(CADT, text="Course code :",bg="#000000",fg="white",font=("Calibri", 18), borderwidth=0).place(x=90,y=230)

global CCDTN
CCDTN = StringVar()

DTCOD_entry=Entry(CADT,bg="#000000",fg="#FFFFFF",textvariable=CCDTN,insertbackground="#00FFFF",font=("Calibri", 15), borderwidth=1)
DTCOD_entry.place(x=230,y=230)

def dtI(e):
    dtimage=PhotoImage(file="images/buttons/cal1.png")
    dtbutton.config(image=dtimage)
    dtbutton.image=dtimage
def Dti(e):
    dtimage=PhotoImage(file="images/buttons/cal.png")
    dtbutton.config(image=dtimage)
    dtbutton.image=dtimage
dtimage=PhotoImage(file="images/buttons/cal.png")
dtbutton = tk.Button(tab7, image=dtimage,command=lambda:[quit_loop_tab7(),Search_cdet()],bg="#000000",activebackground='#000000', borderwidth=0)
dtbutton.bind("<Enter>", dtI)
dtbutton.bind("<Leave>", Dti)

dtbutton.place(x=300,y=475)

t7imagei=PhotoImage(file="images/icons/attend.png")
t7buttoni = tk.Button(tab7, image=t6imagei,bg="black",activebackground='black', borderwidth=0)
t7buttoni.place(x=300,y=100)

###########################################################################################################################################################

######################functions of tab 8 #######################################################################################

def myEntry():
    global myp
    myp=[]
    with open('Teachers/'+userName+'/'+userName, 'r+') as filehandle:
        for line in filehandle:
            myp.append(line)
    newmyp = [x[:-1] for x in myp]

    uEntry=Entry(Ftab8, textvariable=usname,font=("Calibri", 16),fg="#24f2e8",bg="#000000", borderwidth=1,insertbackground="#24f2e8")
    uEntry.place(x=50,y=60)
    uEntry.delete(0, END)
    uEntry.insert(0,newmyp[0])
    eEntry=Entry(Ftab8, textvariable=emid,font=("Calibri", 16),fg="#24f2e8",bg="#000000", borderwidth=1,insertbackground="#24f2e8")
    eEntry.place(x=50,y=140)
    eEntry.delete(0, END)
    eEntry.insert(0,newmyp[1])
    pEntry=Entry(Ftab8, textvariable=psword,show="*",font=("Calibri", 16),fg="#24f2e8",bg="#000000", borderwidth=1,insertbackground="#24f2e8")
    pEntry.place(x=50,y=220)
    pEntry.delete(0, END)
    pEntry.insert(0,newmyp[2])
    cEntry=Entry(Ftab8, textvariable=cmword,show="*",font=("Calibri", 16),fg="#24f2e8",bg="#000000", borderwidth=1,insertbackground="#24f2e8")
    cEntry.place(x=50,y=300)
    cEntry.delete(0, END)
    cEntry.insert(0,newmyp[2])

def reset():
    Usern = usname.get()
    if os.path.exists('Teachers/'+Usern+'/Reset'+Usern):
        os.remove('Teachers/'+Usern+'/'+Usern)
    rep=[]
    with open('Teachers/'+Usern+'/Reset'+Usern, "r+") as resetfile:
        for line in resetfile:
            rep.append(line)
    newrep = [r[:-1] for r in rep]
    os.rename('Teachers/'+Usern+'/Reset'+Usern , 'Teachers/'+Usern+'/'+newrep[0])
    os.rename('Teachers/'+Usern , 'Teachers/'+newrep[0])
    myEntry()
def registerU():
    global register_success_screen
    register_success_screen = Toplevel(form)
    Usern = usname.get()
    Email = emid.get()
    Passw = psword.get()
    Cpassw = cmword.get()
    if Passw == Cpassw:
        if Usern != userName:
            if os.path.exists('Teachers/'+Usern):
                Label(register_success_screen, text="This Username alredy registered ", fg="green", font=("calibri", 20)).grid(row=7,column=2,padx=150, pady=30)
            else:
                os.rename('Teachers/'+userName+'/'+userName , 'Teachers/'+userName+'/Reset'+Usern)
                os.rename('Teachers/'+userName , 'Teachers/'+Usern)
                file = open('Teachers/'+Usern+'/'+Usern, "w")
                file.write(Usern + "\n")
                file.write(Email + "\n")
                file.write(Passw + "\n")
                file.close()
                Label(register_success_screen, text="Updated Successfully", fg="green", font=("calibri", 20)).grid(row=7,column=2,padx=150, pady=30)
        else:
            if os.path.exists('Teachers/'+Usern):
                os.rename('Teachers/'+Usern+'/'+Usern , 'Teachers/'+Usern+'/Reset'+Usern)
                file = open('Teachers/'+Usern+'/'+Usern, "w")
                file.write(Usern + "\n")
                file.write(Email + "\n")
                file.write(Passw + "\n")
                file.close()
                Label(register_success_screen, text="Updated Successfully", fg="green", font=("calibri", 20)).grid(row=7,column=2,padx=150, pady=30)
    else:
        Label(register_success_screen, text="Password and Confirm password must be same", fg="green", font=("calibri", 20)).grid(row=7,column=2,padx=150, pady=30)
        
def removeU():
    shutil.rmtree(r'Teachers/'+userName)
    form.destroy()
def logout():
    form.destroy()

def UploadAction():
    global Filename
    Filename = filedialog.askopenfilename(title="Choose a file",filetypes=[('image files', ('.png', '.jpg'))])
    if (pathlib.Path(Filename).suffix) != '.png':
        im1 = Image.open(r''+Filename)
        im = im1.resize((100, 111))
        im.save(r'Teachers/'+userName+'/user.png')

    else:
        shutil.copy(filename,'Teachers/'+userName+'/user.png')
    profileP()

def profileP():
    global Pimg
    if os.path.exists('Teachers/'+userName+'/user.png'):
        Pimg=PhotoImage(file='Teachers/'+userName+'/user.png')
        buttoni = tk.Button(tab8, text='Open', command=UploadAction, image=Pimg,bg="#ffffff", borderwidth=0).place(x=760,y=60)
    else:
        Pimg=PhotoImage(file='images/icons/user.png')
        buttoni = tk.Button(tab8, text='Open', command=UploadAction, image=Pimg,bg="#ffffff", borderwidth=0).place(x=760,y=60)

############################################### tab8 start #########################################################################################################

filename9 = PhotoImage(file = "images/bg/tab8bg.png")
background_label = Label(tab8, image=filename9)
background_label.place(x=0, y=0, relheight=1)

global usname
global psword
global emid
global cmword
#file = open('Users/'+uname, "r")

Ftab8=Frame(tab8,bg="#000000")
Ftab8.place(x=1052,y=0,height=710,width=410)

usname = StringVar()
idno = StringVar()
psword = StringVar()
emid = StringVar()
cmword = StringVar()

#Label(Ftab8, text="Your profile to view and update", height="2",fg="#ff3300",bg="#000000",font=("Calibri", 20)).place(x=150,y=10)

Label(Ftab8, text="Username:",fg="white",bg="#000000",font=("Calibri", 16), borderwidth=0).place(x=50,y=30)
Label(Ftab8, text="Email ID:",fg="white",bg="#000000",font=("Calibri", 16), borderwidth=0).place(x=50,y=110)
Label(Ftab8, text="Password:",fg="white",bg="#000000",font=("Calibri", 16), borderwidth=0).place(x=50,y=190)
Label(Ftab8, text="Confirm Password:",fg="white",bg="#000000",font=("Calibri", 16), borderwidth=0).place(x=50,y=270)

profileP()
myEntry()

def RI(e):
    imageR=PhotoImage(file="images/buttons/reset2.png")
    buttonR.config(image=imageR)
    buttonR.image=imageR
def ri(e):
    imageR=PhotoImage(file="images/buttons/reset1.png")
    buttonR.config(image=imageR)
    buttonR.image=imageR
imageR=PhotoImage(file="images/buttons/reset1.png")
buttonR = tk.Button(Ftab8, image=imageR,command=reset,bg="#000000", borderwidth=0)
buttonR.bind("<Enter>", RI)
buttonR.bind("<Leave>", ri)

def UI(e):
    imageU=PhotoImage(file="images/buttons/submit2.png")
    buttonU.config(image=imageU)
    buttonU.image=imageU
def ui(e):
    imageU=PhotoImage(file="images/buttons/submit1.png")
    buttonU.config(image=imageU)
    buttonU.image=imageU
imageU=PhotoImage(file="images/buttons/submit1.png")
buttonU = tk.Button(Ftab8, image=imageU,command=registerU,bg="#000000", borderwidth=0)
buttonU.bind("<Enter>", UI)
buttonU.bind("<Leave>", ui)

def DI(e):
    imageD=PhotoImage(file="images/buttons/remove2.png")
    buttonD.config(image=imageD)
    buttonD.image=imageD
def di(e):
    imageD=PhotoImage(file="images/buttons/remove1.png")
    buttonD.config(image=imageD)
    buttonD.image=imageD
imageD=PhotoImage(file="images/buttons/remove1.png")
buttonD = tk.Button(Ftab8, image=imageD,command=removeU,bg="#000000", borderwidth=0)
buttonD.bind("<Enter>",DI)
buttonD.bind("<Leave>",di)

def OI(e):
    imageO=PhotoImage(file="images/buttons/logout2.png")
    buttonO.config(image=imageO)
    buttonO.image=imageO
def Oi(e):
    imageO=PhotoImage(file="images/buttons/logout1.png")
    buttonO.config(image=imageO)
    buttonO.image=imageO
imageO=PhotoImage(file="images/buttons/logout1.png")
buttonO = tk.Button(Ftab8, image=imageO,command=logout,bg="#000000", borderwidth=0)
buttonO.bind("<Enter>", OI)
buttonO.bind("<Leave>", Oi)

buttonR.place(x=80,y=430)
buttonU.place(x=80,y=490)
buttonD.place(x=80,y=550)
buttonO.place(x=80,y=610)

############################ Tab8 END   ################################################################################################################

tab_parent.pack(expand=1, fill='both')

form.mainloop()
```
